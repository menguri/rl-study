# 0. DQN 전체 코드

DQN-based 알고리즘들은 DQN 코드를 기반으로 구현하기 때문에 아래 DQN 전체 코드를 먼저 실행시킨 후 실습을 진행

```python
# 필요 라이브러리 로드

import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import gym
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
# 지저분한 deprecation warning 문구 제거

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# hyperparameter 정의
config = dict(
    env_name            = "Acrobot-v1", # 환경
    batch               = 128,          # 배치 크기
    buffer_size         = 10000,        # 버퍼의 최대 크기(transition 단위)
    lr                  = 0.001,        # 학습률
    gamma               = 0.99,         # discount(할인율)
    epsilon_init        = 0.1,          # 초기 epsilon(=초기 탐험 비율)
    epsilon_min         = 0.001,        # 최소 epsilon(=최종 탐험 비율)
    epsilon_decay       = 0.995,        # epsilon 감소 비율(매 episode마다 epsilon에 곱해짐)
    n_step              = 50000,        # 학습 횟수 (time step 단위)
    n_train_start       = 8000,         # 초기 버퍼 채우기(time step 단위)
    target_update_freq  = 100,          # target Q의 업데이트 주기(time step 단위)
    test_freq           = 100,          # test 주기(time step 단위)
)

# 환경 및 테스트 환경 정의
env      = gym.make(config["env_name"])
test_env = gym.make(config["env_name"])
dState = env.observation_space.shape[0] # 환경의 state_space 크기
dAction = env.action_space.n # 환경의 discrete action 개수

# Q 네트워크 클래스 정의
class Qnet(nn.Module):
    def __init__(self, dState, dAction):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(in_features=dState, out_features=128),    # 입력층
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=128),       # 은닉층
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=dAction),   # 출력층
        )
        # 신경망 모델 정의 총 3개의 레이어로 구성
        # 입력(state 차원) -> 입력층(128) -> 은닉층(128) -> 출력층(action 차원)

    def forward(self, x): # 네트워크 처리 과정 정의
        return self.layers(x) # self.layers 신경망을 통과한 후 출력을 반환


# Q 및 target Q 네트워크 정의
Q = Qnet(dState=dState, dAction=dAction).to(DEVICE)
targetQ = Qnet(dState=dState, dAction=dAction).to(DEVICE) # target Q 정의
targetQ.load_state_dict(Q.state_dict())
optimizerQ = optim.Adam(Q.parameters(), lr=config["lr"])

# replay buffer 정의
replay_buffer = deque(maxlen=config["buffer_size"])

gamma = config["gamma"]
epsilon = config["epsilon_init"]

def getAction(state, dAction, epsilon, Q):
    with torch.no_grad(): # 본 함수는 역전파를 위한 gradient 계산이 필요 없으므로 gradient 계산을 제외시켜 연산량 줄임
        if (random.random() > epsilon): # exploitation(활용)
            state = torch.from_numpy(state).float().to(DEVICE)
            # Q 네트워크가 device에 존재하므로 입력인 state도 동일한 device로 이동시켜줌
            action = Q(state).argmax(dim=-1).detach().cpu().numpy()
            # state에서 모든 action에 대해 Q를 계산하고, 가장 높은 Q를 가진 action 선택
        else: # exploration(탐험)
            action = np.random.randint(low=0, high=dAction)
            # random하게 action 선택
    return action

def test(env, Q, dAction, epsilon):
    with torch.no_grad():
        score = 0.0
        state = env.reset()
        while True:
            action = getAction(state, dAction, epsilon, Q)
            nextState, reward, done, _info = env.step(action)
            score += reward
            state = nextState
            if done:
                break
    return score


# 학습 전, 샘플 수집
state = env.reset()
for t in range(1, config["n_train_start"]+1):
    action = getAction(state, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    transition = (state, action, reward, nextState, done)
    replay_buffer.append(transition)
    state = nextState
    if done:
        state = env.reset()

# 메인 학습 루프
score = 0.0
scores = []
test_scores = []
losses = []
qs = []
# score 및 각종 성능 계산을 위한 변수 정의

state = env.reset() # (dState)
# 환경 초기화 및 초기 상태 획득
for t in range(1, config["n_step"]+1):
    # --- 환경 상호작용 ---
    action = getAction(state, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    transition = (state, action, reward, nextState, done)
    replay_buffer.append(transition)

    score += reward
    state = nextState
    if done: # episode 종료 시
        state = env.reset()
        scores.append(score)
        score = 0.0
        # --- epsilon 업데이트 ---
        epsilon = max(config["epsilon_min"], epsilon*config["epsilon_decay"])

    # --- batch 샘플링 ---
    transitions = random.sample(replay_buffer, config["batch"])
    batch = []
    for item in zip(*transitions):
        item = torch.from_numpy(np.stack(item)).float().to(DEVICE)
        batch.append(item)
    states, actions, rewards, nextStates, dones = batch # (dBatch, -)
    actions = actions.unsqueeze(dim=-1).long()
    rewards = rewards.unsqueeze(dim=-1)
    dones = dones.unsqueeze(dim=-1)
    # 배치 샘플 준비

    # --- loss 계산 ---
    # estimate
    estimateQs = Q(states).gather(dim=1, index=actions) # (dBatch, 1)
    # Q(s, a)

    # target
    with torch.no_grad():
        nextTargetQs = targetQ(nextStates).max(dim=-1, keepdim=True).values # (dBatch, 1)
        # max_a Q'(s_t+1, a)
        targetQs = rewards + (1 - dones) * gamma * nextTargetQs # (dBatch, 1)
        # r + gamma * max_a Q'(s_t+1, a)

    loss = F.mse_loss(targetQs.detach(), estimateQs) # scalar
    # MSE{ targetQ, Q }

    # --- 오차 역전파 및 매개변수 업데이트 ---
    optimizerQ.zero_grad()
    loss.backward()
    optimizerQ.step()
    losses.append(loss.detach().cpu().item())
    qs.append(estimateQs.detach().mean().cpu().item())

    # --- target Q 네트워크 업데이트 ---
    if t % config["target_update_freq"] == 0:
        targetQ.load_state_dict(Q.state_dict())
        # 정해진 주기마다 targetQ 를 Q 네트워크로 업데이트

    # --- test score 계산 ---
    if t % config["test_freq"] == 0:
        test_score = test(test_env, Q, dAction, epsilon=config["epsilon_min"])
        test_scores.append(test_score)
        print(f'[time:{t}]test score: {test_score}, train loss: {losses[-1]:.3f}, epsilon: {epsilon:.3f}')
        # 정해진 주기마다 현재 모델로 환경과 한 episode 상호작용후 score 계산
```

# 1. Double DQN

Double DQN은 DQN코드의 loss 부분에서 target Q를 계산하는 부분만이 수정됨

```python
# estimate Q 계산
## DQN과 동일

################################### 코드 작성 ###################################

estimateQs = Q(states).gather(dim=1, index=actions)
# estimateQs: batch 상태(states)와 batch 행동(actions)으로 Q를 계산

#################################################################################
print("estimateQs의 형태:", estimateQs.shape)
print("estimateQs:", estimateQs)
```

```python
# target Q 계산

################################### 코드 작성 ###################################

nextActions = Q(nextStates).argmax(dim=-1, keepdim=True)
# nextActions: online network(Q)로 다음 상태에서의 가치를 계산하고, 가장 가치가 큰 행동을 선택
## argmax_a' Q(s', a')

nextTargetQs = targetQ(nextStates).gather(dim=-1, index=nextActions)
# nextTargetQs: online network(Q)로 얻어진 행동(nextActions)의 가치를
# target network(targetQ)를 활용하여 평가
## Q'(s', argmax_a' Q(s', a'))

targetQs = rewards + (1 - dones) * gamma * nextTargetQs
# targetQs: 1-step TD target 방식으로 target Q 계산
## r + gamma * Q'(s', argmax_a' Q(s', a'))

#################################################################################
print("online network Q(nextStates[0]):", Q(nextStates[0]))
print("online network Q(nextStates[0])에서 가장 큰 Q:", Q(nextStates[0]).max().cpu().item())
print("online network Q(nextStates[0])에서 가장 큰 Q의 행동:", Q(nextStates).argmax(dim=-1, keepdim=True)[0].cpu().item())
print()

print("target network Q'(nextStates[0]):", targetQ(nextStates[0]))
print("target network Q'(nextStates[0])에서 가장 큰 Q:", targetQ(nextStates[0]).max().cpu().item())
print("target network Q'(nextStates[0])에서 가장 큰 Q의 행동:", targetQ(nextStates).argmax(dim=-1, keepdim=True)[0].cpu().item())
print()

# online network로 다음 행동 선택 후 target network로 평가
print("online network에서 행동 선택:", nextActions[0].cpu().item())
print("선택된 행동의 Q를 target network로 평가:", nextTargetQs[0].cpu().item())
```

```python
# MSE loss 계산

################################### 코드 작성 ###################################

loss = F.mse_loss(targetQs.detach(), estimateQs)
# loss: MSE loss 계산 - DQN과 동일

#################################################################################

print("MES loss:", loss)
```

### 1-2. Double DQN 전체 코드

```python
# 약 5~6분 소요

# 필요 라이브러리 로드
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import gym
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
# 지저분한 deprecation warning 문구 제거

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# hyperparameter 정의
config = dict(
    env_name            = "Acrobot-v1", # 환경
    batch               = 128,          # 배치 크기
    buffer_size         = 10000,        # 버퍼의 최대 크기(transition 단위)
    lr                  = 0.001,        # 학습률
    gamma               = 0.99,         # discount(할인율)
    epsilon_init        = 0.1,          # 초기 epsilon(=초기 탐험 비율)
    epsilon_min         = 0.001,        # 최소 epsilon(=최종 탐험 비율)
    epsilon_decay       = 0.995,        # epsilon 감소 비율(매 episode마다 epsilon에 곱해짐)
    n_step              = 50000,        # 학습 횟수 (time step 단위)
    n_train_start       = 8000,         # 초기 버퍼 채우기(time step 단위)
    target_update_freq  = 100,          # target Q의 업데이트 주기(time step 단위)
    test_freq           = 100,          # test 주기(time step 단위)
)

# 환경 및 테스트 환경 정의
env      = gym.make(config["env_name"])
test_env = gym.make(config["env_name"])
dState = env.observation_space.shape[0] # 환경의 state_space 크기
dAction = env.action_space.n # 환경의 discrete action 개수

# Q 네트워크 클래스 정의
class Qnet(nn.Module):
    def __init__(self, dState, dAction):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(in_features=dState, out_features=128),    # 입력층
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=128),       # 은닉층
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=dAction),   # 출력층
        )
        # 신경망 모델 정의 총 3개의 레이어로 구성
        # 입력(state 차원) -> 입력층(128) -> 은닉층(128) -> 출력층(action 차원)

    def forward(self, x): # 네트워크 처리 과정 정의
        return self.layers(x) # self.layers 신경망을 통과한 후 출력을 반환


# Q 및 target Q 네트워크 정의
Q = Qnet(dState=dState, dAction=dAction).to(DEVICE)
targetQ = Qnet(dState=dState, dAction=dAction).to(DEVICE) # target Q 정의
targetQ.load_state_dict(Q.state_dict())
optimizerQ = optim.Adam(Q.parameters(), lr=config["lr"])

# replay buffer 정의
replay_buffer = deque(maxlen=config["buffer_size"])

gamma = config["gamma"]
epsilon = config["epsilon_init"]

def getAction(state, dAction, epsilon, Q):
    with torch.no_grad(): # 본 함수는 역전파를 위한 gradient 계산이 필요 없으므로 gradient 계산을 제외시켜 연산량 줄임
        if (random.random() > epsilon): # exploitation(활용)
            state = torch.from_numpy(state).float().to(DEVICE)
            # Q 네트워크가 device에 존재하므로 입력인 state도 동일한 device로 이동시켜줌
            action = Q(state).argmax(dim=-1).detach().cpu().numpy()
            # state에서 모든 action에 대해 Q를 계산하고, 가장 높은 Q를 가진 action 선택
        else: # exploration(탐험)
            action = np.random.randint(low=0, high=dAction)
            # random하게 action 선택
    return action

def test(env, Q, dAction, epsilon):
    with torch.no_grad():
        score = 0.0
        state = env.reset()
        while True:
            action = getAction(state, dAction, epsilon, Q)
            nextState, reward, done, _info = env.step(action)
            score += reward
            state = nextState
            if done:
                break
    return score


# 학습 전, 샘플 수집
state = env.reset()
for t in range(1, config["n_train_start"]+1):
    action = getAction(state, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    transition = (state, action, reward, nextState, done)
    replay_buffer.append(transition)
    state = nextState
    if done:
        state = env.reset()

# 메인 학습 루프
score = 0.0
scores1 = []
test_scores1 = []
losses1 = []
qs1 = []
# score 및 각종 성능 계산을 위한 변수 정의

state = env.reset() # (dState)
# 환경 초기화 및 초기 상태 획득
for t in range(1, config["n_step"]+1):
    # --- 환경 상호작용 ---
    action = getAction(state, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    transition = (state, action, reward, nextState, done)
    replay_buffer.append(transition)

    score += reward
    state = nextState
    if done: # episode 종료 시
        state = env.reset()
        scores1.append(score)
        score = 0.0
        # --- epsilon 업데이트 ---
        epsilon = max(config["epsilon_min"], epsilon*config["epsilon_decay"])

    # --- batch 샘플링 ---
    transitions = random.sample(replay_buffer, config["batch"])
    batch = []
    for item in zip(*transitions):
        item = torch.from_numpy(np.stack(item)).float().to(DEVICE)
        batch.append(item)
    states, actions, rewards, nextStates, dones = batch # (dBatch, -)
    actions = actions.unsqueeze(dim=-1).long()
    rewards = rewards.unsqueeze(dim=-1)
    dones = dones.unsqueeze(dim=-1)
    # 배치 샘플 준비

    # --- loss 게산 ---

    # estimate
    estimateQs = Q(states).gather(dim=1, index=actions) # (dBatch, 1)
    # Q(s, a)

    # 1-1. loss(오차) 계산
    with torch.no_grad():
        nextActions = Q(nextStates).argmax(dim=-1, keepdim=True) # (dBatch, 1)
        # argmax_a' Q(s', a')
        nextTargetQs = targetQ(nextStates).gather(dim=-1, index=nextActions) # (dBatch, 1)
        # Q'(s', argmax_a' Q(s', a'))
        targetQs = rewards + (1 - dones) * gamma * nextTargetQs # (dBatch, 1)
        # r + gamma * Q'(s', argmax_a' Q(s', a'))

    loss = F.mse_loss(targetQs.detach(), estimateQs) # scalar
    # MSE{ targetQ, Q }

    # --- 오차 역전파 및 매개변수 업데이트 ---
    optimizerQ.zero_grad()
    loss.backward()
    optimizerQ.step()
    losses1.append(loss.detach().cpu().item())
    qs1.append(estimateQs.detach().mean().cpu().item())

    # --- target Q 네트워크 업데이트 ---
    if t % config["target_update_freq"] == 0:
        targetQ.load_state_dict(Q.state_dict())
        # 정해진 주기마다 targetQ 를 Q 네트워크로 업데이트

    # --- test score 계산 ---
    if t % config["test_freq"] == 0:
        test_score = test(test_env, Q, dAction, epsilon=config["epsilon_min"])
        test_scores1.append(test_score)
        print(f'[time:{t}]test score: {test_score}, train loss: {losses1[-1]:.3f}, epsilon: {epsilon:.3f}')
        # 정해진 주기마다 현재 모델로 환경과 한 episode 상호작용후 score 계산
```

# 2. Dueling DQN

Dueling DQN은 DQN에서 Q 네트워크 정의하는 부분만이 수정됨

### 2-1. Q 네트워크 정의

```python
# Q 네트워크 클래스 정의

################################### 코드 작성 ###################################

class Qnet(nn.Module):
    def __init__(self, dState, dAction):
        super().__init__()
        # nn.Module의 초기화
        self.featureLayer = nn.Sequential(
            nn.Linear(in_features=dState, out_features=128),
            nn.ReLU(),
        )
        # self.featureLayer: 공용 네트워크 정의
        ## dState(상태 차원) -> 128 차원
        ## 활성화 함수는 ReLU 사용
        self.valueLayer = nn.Sequential(
            nn.Linear(in_features=128, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=1),
        )
        # self.valueLayer: value 네트워크 정의
        ## 128 차원 -> 128 차원 -> 1 차원
        ## 활성화 함수는 ReLU 사용
        self.advantageLayer = nn.Sequential(
            nn.Linear(in_features=128, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=dAction),
        )
        # self.advantageLayer: advantage 네트워크 정의
        ## 128 차원 -> 128 차원 -> dAction(행동 차원)
        ## 활성화 함수는 ReLU 사용

    def forward(self, x):                        # (dBatch, dState)
    # forward 메서드: 상태를 입력으로 받아 모든 행동 각각에 대한 Q 반환
        feature = self.featureLayer(x)           # (dBatch, dHidden)
        # feature: 공통 네트워크 통과된 특징벡터 저장
        value = self.valueLayer(feature)         # (dBatch, 1)
        # value: value 네트워크 통과된 value 저장
        advantage = self.advantageLayer(feature) # (dBatch, dAction)
        # advantage: advantage 네트워크 통과된 advantage 저장
        Q = value + (advantage - advantage.mean(dim=-1, keepdim=True)) # (dBatch, dAction)
        # Q: value와 advantage를 사용하여 Q 계산
        return Q # (dBatch, dAction)

#################################################################################
```

```python
# Q 및 target Q 네트워크 정의

Q = Qnet(dState=dState, dAction=dAction).to(DEVICE)
# Q 네트워크 선언(online network) 및 device 이동(GPU or CPU)
targetQ = Qnet(dState=dState, dAction=dAction).to(DEVICE) # target Q 정의
# target Q 네트워크 선언(target network) 및 device 이동(GPU or CPU)
```

```python
# advantage 계산

advantage = Q.advantageLayer(feature) # (dBatch, dAction)
print("feature 형태:", feature.shape)
print("advantage의 형태:", advantage.shape)
print(advantage)
```

```python
# Q 계산

mean = advantage.mean(dim=-1, keepdim=True)
# advantage의 평균 계산
print("advantage mean의 형태:", mean.shape)

Q = value + (advantage - mean) # (dBatch, dAction)
# value와 advantage로 Q 계산
print("Q의 형태:", Q.shape)
print(Q)
```

### 2-2. Dueling DQN의 전체 코드

```python
# 약 5~6분 소요

# 필요 라이브러리 로드
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import gym
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
# 지저분한 deprecation warning 문구 제거

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# hyperparameter 정의
config = dict(
    env_name            = "Acrobot-v1", # 환경
    batch               = 128,          # 배치 크기
    buffer_size         = 10000,        # 버퍼의 최대 크기(transition 단위)
    lr                  = 0.001,        # 학습률
    gamma               = 0.99,         # discount(할인율)
    epsilon_init        = 0.1,          # 초기 epsilon(=초기 탐험 비율)
    epsilon_min         = 0.001,        # 최소 epsilon(=최종 탐험 비율)
    epsilon_decay       = 0.995,        # epsilon 감소 비율(매 episode마다 epsilon에 곱해짐)
    n_step              = 50000,        # 학습 횟수 (time step 단위)
    n_train_start       = 8000,         # 초기 버퍼 채우기(time step 단위)
    target_update_freq  = 100,          # target Q의 업데이트 주기(time step 단위)
    test_freq           = 100,          # test 주기(time step 단위)
)

# 환경 및 테스트 환경 정의
env      = gym.make(config["env_name"])
test_env = gym.make(config["env_name"])
dState = env.observation_space.shape[0] # 환경의 state_space 크기
dAction = env.action_space.n # 환경의 discrete action 개수

# 2-1. Q 네트워크 클래스 정의
class Qnet(nn.Module):
    def __init__(self, dState, dAction):
        super().__init__()
        self.featureLayer = nn.Sequential( # 공통 네트워크
            nn.Linear(in_features=dState, out_features=128),
            nn.ReLU(),
        )
        self.valueLayer = nn.Sequential( # value 네트워크
            nn.Linear(in_features=128, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=1),
        )
        self.advantageLayer = nn.Sequential( # advantage 네트워크
            nn.Linear(in_features=128, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=dAction),
        )

    def forward(self, x):     # (dBatch, dState)
        feature = self.featureLayer(x) # (dBatch, dHidden)
        value = self.valueLayer(feature) # (dBatch, 1)
        advantage = self.advantageLayer(feature) # (dBatch, dAction)
        # 공통 네트워크까지 계산된 feature를 각각 분리된 value, advantage 네트워크에 넣어줌
        Q = value + (advantage - advantage.mean(dim=-1, keepdim=True)) # (dBatch, dAction)
        # value와 advantage를 사용하여 Q 계산
        return Q # (dBatch, dAction)

# Q 및 target Q 네트워크 정의
Q = Qnet(dState=dState, dAction=dAction).to(DEVICE)
targetQ = Qnet(dState=dState, dAction=dAction).to(DEVICE) # target Q 정의
targetQ.load_state_dict(Q.state_dict())
optimizerQ = optim.Adam(Q.parameters(), lr=config["lr"])

# replay buffer 정의
replay_buffer = deque(maxlen=config["buffer_size"])

gamma = config["gamma"]
epsilon = config["epsilon_init"]

def getAction(state, dAction, epsilon, Q):
    with torch.no_grad(): # 본 함수는 역전파를 위한 gradient 계산이 필요 없으므로 gradient 계산을 제외시켜 연산량 줄임
        if (random.random() > epsilon): # exploitation(활용)
            state = torch.from_numpy(state).float().to(DEVICE)
            # Q 네트워크가 device에 존재하므로 입력인 state도 동일한 device로 이동시켜줌
            action = Q(state).argmax(dim=-1).detach().cpu().numpy()
            # state에서 모든 action에 대해 Q를 계산하고, 가장 높은 Q를 가진 action 선택
        else: # exploration(탐험)
            action = np.random.randint(low=0, high=dAction)
            # random하게 action 선택
    return action

def test(env, Q, dAction, epsilon):
    with torch.no_grad():
        score = 0.0
        state = env.reset()
        while True:
            action = getAction(state, dAction, epsilon, Q)
            nextState, reward, done, _info = env.step(action)
            score += reward
            state = nextState
            if done:
                break
    return score


# 학습 전, 샘플 수집
state = env.reset()
for t in range(1, config["n_train_start"]+1):
    action = getAction(state, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    transition = (state, action, reward, nextState, done)
    replay_buffer.append(transition)
    state = nextState
    if done:
        state = env.reset()

# 메인 학습 루프
score = 0.0
scores2 = []
test_scores2 = []
losses2 = []
qs2 = []
# score 및 각종 성능 계산을 위한 변수 정의

state = env.reset() # (dState)
# 환경 초기화 및 초기 상태 획득
for t in range(1, config["n_step"]+1):
    # --- 환경 상호작용 ---
    action = getAction(state, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    transition = (state, action, reward, nextState, done)
    replay_buffer.append(transition)

    score += reward
    state = nextState
    if done: # episode 종료 시
        state = env.reset()
        scores2.append(score)
        score = 0.0
        # --- epsilon 업데이트 ---
        epsilon = max(config["epsilon_min"], epsilon*config["epsilon_decay"])

    # --- batch 샘플링 ---
    transitions = random.sample(replay_buffer, config["batch"])
    batch = []
    for item in zip(*transitions):
        item = torch.from_numpy(np.stack(item)).float().to(DEVICE)
        batch.append(item)
    states, actions, rewards, nextStates, dones = batch # (dBatch, -)
    actions = actions.unsqueeze(dim=-1).long()
    rewards = rewards.unsqueeze(dim=-1)
    dones = dones.unsqueeze(dim=-1)
    # 배치 샘플 준비

    # --- loss 계산 ---
    # estimate
    estimateQs = Q(states).gather(dim=1, index=actions) # (dBatch, 1)
    # Q(s, a)

    # target
    with torch.no_grad():
        nextTargetQs = targetQ(nextStates).max(dim=-1, keepdim=True).values # (dBatch, 1)
        # max_a Q'(s', a)
        targetQs = rewards + (1 - dones) * gamma * nextTargetQs # (dBatch, 1)
        # r + gamma * max_a Q'(s', a)

    loss = F.mse_loss(targetQs.detach(), estimateQs) # scalar
    # MSE{ targetQ, Q }

    # --- 오차 역전파 및 매개변수 업데이트 ---
    optimizerQ.zero_grad()
    loss.backward()
    optimizerQ.step()
    losses2.append(loss.detach().cpu().item())
    qs2.append(estimateQs.detach().mean().cpu().item())

    # --- target Q 네트워크 업데이트 ---
    if t % config["target_update_freq"] == 0:
        targetQ.load_state_dict(Q.state_dict())
        # 정해진 주기마다 targetQ 를 Q 네트워크로 업데이트

    # --- test score 계산 ---
    if t % config["test_freq"] == 0:
        test_score = test(test_env, Q, dAction, epsilon=config["epsilon_min"])
        test_scores2.append(test_score)
        print(f'[time:{t}]test score: {test_score}, train loss: {losses2[-1]:.3f}, epsilon: {epsilon:.3f}')
        # 정해진 주기마다 현재 모델로 환경과 한 episode 상호작용후 score 계산
```

# 3. DRQN(Deep Recurrent Q-Network)

행동(action)을 선택할때 이전 상태(state)만을 고려하는 DQN과 달리, 이전 경로(trajectory)를 고려하므로 코드의 많은 부분이 수정됨

### 3-1. hyperparameter 정의

```python
# hyperparameter 정의

config = dict(
    env_name            = "Acrobot-v1",
    batch               = 8,
    buffer_size         = 100, # 버퍼의 최대 크기(episode 단위)
    lr                  = 0.001, # 학습률
    gamma               = 0.99, # discount(할인율)
    epsilon_init        = 0.1, # 초기 epsilon
    epsilon_min         = 0.001, # 최소 epsilon
    epsilon_decay       = 0.995,
    n_step              = 50000, # 학습 횟수
    n_train_start       = 8000,
    target_update_freq  = 100,  # target Q의 업데이트 주기(time step 단위)
    test_freq           = 100, # test 주기(time step 단위)
    max_sample_len      = 500, # sampling 시 최대 transition 시퀀스 길이
    sample_length       = 10, # sampling 시 시퀀스 길이
)

epsilon = config["epsilon_init"]
# 학습시 사용하는 epsilon 을 따로 선언
```

### 3-2. Q 네트워크 정의

```python
# Q 네트워크 클래스 정의

################################### 코드 작성 ###################################

class Qnet(nn.Module):
    def __init__(self, dState, dAction):
        super().__init__()
        self.dHidden = 128
        self.inputLayer = nn.Linear(in_features=dState, out_features=self.dHidden) # 입력층
        # self.inputLayer: 입력층 정의
        ## dState(상태 차원) -> 128 차원
        self.lstm = nn.LSTM(input_size=self.dHidden, hidden_size=self.dHidden, batch_first=True) # 은닉층=LSTM
        # self.lstm: nn.LSTM으로 은닉층(순환 신경망) 정의
        ## 128 차원 -> 128 차원
        self.outputLayer = nn.Linear(in_features=self.dHidden, out_features=dAction) # 출력층
        # self.outputLayer: 출력층 정의
        ## 128 차원 -> dAction(행동 개수)

    def init_state(self, dBatch=1):
    # init_state 메서드: 초기갑 0으로 설정된 장/단기 메모리 반환
    ## 환경과 상호작용 시 batch 크기=1 이며, 학습시 batch 단위이고 batch 각각의 샘플마다 서로 다른 장/단기 메모리 사용
        h = torch.zeros([1, dBatch, self.dHidden]).to(DEVICE)
        # h: 단기 메모리 정의
        c = torch.zeros([1, dBatch, self.dHidden]).to(DEVICE)
        # c: 장기 메모리 정의
        return h, c

    def forward(self, x, h, c): # (dBatch, dSeq, dState), (1, dBatch, dHidden), (1, dBatch, dHidden)
    # 상태(x), 이전 단기 메모리(h), 이전 장기 메모리(c)를 입력으로 받아 Q를 계산하는 네트워크 처리 과정 정의
        x = F.relu(self.inputLayer(x)) # (dBatch, dSeq, dHidden)
        # x: 입력층 및 활성화함수(ReLU 통과
        x, (newH, newC) = self.lstm(x, (h, c)) # (dBatch, dSeq, dHidden), (1, dBatch, dHidden), (1, dBatch, dHidden)
        # x, (newH, newC): 은닉층 통과 및 time step별 장/단기 메모리 값 저장
        ## 은닉층=LSTM을 통과, 순환신경망이므로 내부적으로 sequnce 데이터가 순환하여 통과됨
        x = self.outputLayer(x) # (dBatch, dSeq, dAction)
        # x: 최종 출력층 통과하여 모든 행동 별 Q값 저장
        return x, newH, newC # (dBatch, dSeq, dState), (1, dBatch, dHidden), (1, dBatch, dHidden)
        # 출력값 및 장/단기 메모리를 함께 반환

#################################################################################
```

```python
# Q 및 target Q 네트워크 정의

Q = Qnet(dState=dState, dAction=dAction).to(DEVICE)
# Q 네트워크 선언(online network) 및 device 이동(GPU or CPU)
targetQ = Qnet(dState=dState, dAction=dAction).to(DEVICE) # target Q 정의
# target Q 네트워크 선언(target network) 및 device 이동(GPU or CPU)
```

```python
# Q 구성 확인

print(Q)
# DQN과 유사하게 입력층, 은닉층, 출력층 각각 1층으로 구성되지만
# 은닉층이 순환 신경망인 LSTM으로 구성됨
# 활성화 함수는 모델안에 포함시키지 않았으므로, forward 계산 시 직접 적용
```

```python
# 초기 장/단기 메모리 생성

h, c = Q.init_state(config["batch"])
# h, c: 초기 장/단기 메모리 획득

print("단기 메모리 형태:", h.shape)
print("장기 메모리 형태:", c.shape)
# 장/단기 메모리는 (1, batch크기, 은닉층크기) 형태로 구성

print("단기 메모리 일부:", h[0][0])
print("장기 메모리 일부:", c[0][0])
# 초기 장/단기 메모리는 모두 0으로 설정됨
```

#### exploitation

```python
# exploitation의 경우

################################### 코드 작성 ###################################

state_numpy = env.reset()
# state_numpy: 환경 초기화 및 numpy 형식의 초기 상태 획득
state_torch = torch.from_numpy(state_numpy).float().to(DEVICE)
# state_torch: numpy state를 torch tensor로 변환 및 device 이동
state = state_torch.view(1, 1, -1) # (1, 1, dState)
# state: Q 네트워크에 넣기 위해 batch와 sequence 차원 추가

#################################################################################

# 상태 출력
print("state_numpy 형태:", state_numpy.shape)
print("state_numpy:", state_numpy)
print()

print("state_torch 형태:", state_torch.shape)
print("state_torch:", state_torch)
print()

print("state 형태:", state.shape)
print("state:", state)
```

```python
# Q 네트워크를 통해 가치 계산
## 이때 이전 상태/경로 정보는 장/단기 메모리가 가지고 있음

################################### 코드 작성 ###################################

h, c = Q.init_state()
# h, c: 초기 장/단기 메모리 획득
## 여기서는 예시를 위해 초기값을 사용함

q, newH, newC = Q(state, h, c)
# q, newH, newC: 장/단기 메모리와 상태(state)를 기반으로 모든 행동 각각에 대한 가치 Q 계산

#################################################################################

print("계산된 Q의 형태:", q.shape)
print(q)
```

```python
# 가치를 최대화 하는 행동 선택

################################### 코드 작성 ###################################

action = q.argmax(dim=-1).detach().cpu().item()
# action: 앞서 계산된 q를 통해 가장 큰 가치를 가진 행동 선택

#################################################################################

print(action)
```

#### exploration

```python
# exploration의 경우
## DQN과 동일

action = np.random.randint(low=0, high=dAction)
print(action)
# 랜덤하게 행동을 샘플링
```

```python
# 종합 코드
## 앞서 다룬 exploration, exploitation에 대한 내용을 getAction 함수로 정의

def getAction(state, h, c, dAction, epsilon, Q): # (dState)
    with torch.no_grad():
        state = torch.from_numpy(state).float().to(DEVICE).view(1, 1, -1) # (1, 1, dState)
        q, newH, newC = Q(state, h, c) # (1, 1, dAction), (1, 1, dHidden), (1, 1, dHidden)
        if (random.random() > epsilon): # exploitation
            action = q.argmax(dim=-1).detach().cpu().item()
        else: # exploration
            action = np.random.randint(low=0, high=dAction) # (-)
    return action, newH, newC
```

### 3-5. 학습 전, 샘플 수집

```python
# transition 단위로 샘플을 수집하는 DQN과 달리, episode 단위로 샘플을 수집하고 활용함


################################### 코드 작성 ###################################

h, c = Q.init_state()
# h, c: 초기 장/단기 메모리 초기화
state = env.reset()
# state: 초기 상태 획득
ep = []
# ep: episode의 transition 저장을 위한 임시 변수
for t in range(1, config["n_train_start"]+1):
    action, h, c = getAction(state, h, c, dAction, epsilon, Q)
    # action, h, c: 앞서 정의한 getAction 함수를 활용해 행동 선택
    nextState, reward, done, _info = env.step(action)
    # nextState, reward, done, _info: action으로 환경과 상호작용하여 환경정보 획득
    transition = (state, action, reward, nextState, done)
    # transition: state, action, reward, nextState, done의 튜플 구성
    ep.append(transition)
    # buffer에 transition을 바로 넣지 않고 임시 변수에 저장
    state = nextState
    # state: nextState로 변경

    if done:
    # 환경 종료 시 초기화 및 buffer에 ep 샘플 추가
        state = env.reset()
        h, c = Q.init_state()
        buffer.append(ep)
        ep = []

#################################################################################
```

### 3-6. batch 샘플링

```python
# batch 샘플링
## episode를 랜덤 샘플링 후 다시 시퀀스 길이만큼 transition의 랜덤 샘플링을 수행

################################### 코드 작성 ###################################

episodes = random.sample(buffer, config["batch"])
# episodes: batch 크기(config["batch"]) 만큼 episode를 랜덤 샘플링

#################################################################################
print("episode 개수:", len(episodes))
```

```python
# episode 내에서 시퀀스 길이만큼 랜덤 샘플링

################################### 코드 작성 ###################################

minStep = config["max_sample_len"] # 500
# minStep: 최대 초기 minStep은 config["max_sample_len"]으로 설정
for episode in episodes:
    minStep = min(minStep, len(episode))
# minStep: episodes 내의 모든 episode에 대하여 루프를 돌며 가장 짧은 episode 길이를 계산
## 샘플링 하고자 하는 시퀀스 길이보다 minStep이 더 작은 경우
## 시퀀스 길이로 샘플링이 불가능 하므로 이때는 minStep 길이로 샘플링을 수행함

#################################################################################

print("최소 episode 길이:", minStep)
print("설정해둔 샘플링 시퀀스 길이:", config["sample_length"])
print("본 예시에서 실제 샘플링 길이는", min(minStep, config["sample_length"]), "사용")
```

```python
# episode 별 transition 랜덤 샘플링


samples = []
# samples: 샘플링 된 transition 저장을 위한 빈 리스트 변수
for episode in episodes:
# 샘플링한 episode마다 루프
    if minStep > config["sample_length"]:
        idx = np.random.randint(0, len(episode)-config["sample_length"]+1)
        sample = [np.array(item) for item in zip(*episode[idx:idx+config["sample_length"]])]
        samples.append(sample)
        # 샘플링 시퀀스 길이만큼 transition을 샘플링하여 samples 리스트에 추가
    else:
        idx = np.random.randint(0, len(episode)-minStep+1)
        sample = [np.array(item) for item in zip(*episode[idx:idx+minStep])]
        samples.append(sample)
        # 최소 episode 길이만큼 transition을 샘플링하여 samples 리스트에 추가
```

```python
print("샘플 개수:", len(samples), "시퀀스 길이:", min(minStep, config["sample_length"]))
print()

print("단일 샘플 구성 예시:")
print("상태 시퀀스:")
print(samples[0][0])
print()

print("행동 시퀀스:")
print(samples[0][1])
print()

print("보상 시퀀스:")
print(samples[0][2])
print()

print("다음 상태 시퀀스:")
print(samples[0][3])
print()

print("episode 종료 여부 시퀀스:")
print(samples[0][4])
```

```python
# 학습을 위해 샘플을 batch 형태로 변형

################################### 코드 작성 ###################################

batch = []
# batch: batch 샘플 저장을 위한 빈 리스트 정의
for item in zip(*samples):
    item = torch.from_numpy(np.stack(item)).float().to(DEVICE)
    batch.append(item)
    # 단일 transition 샘플을 stack 하여 batch 리스트에 추가
states, actions, rewards, nextStates, dones = batch

#################################################################################

# (batch 크기, 시퀀스 크기, -)
print("batch의 state 형태:", states.shape)
print("batch의 action 형태:", actions.shape)
print("batch의 reward 형태:", rewards.shape)
print("batch의 nextState 형태:", nextStates.shape)
print("batch의 done 형태:", dones.shape)
```

### 3-7. loss(오차) 계산

```python
# estimate Q 계산
## 장/단기 메모리만 추가되었고, 큰 틀은 DQN과 유사

################################### 코드 작성 ###################################

h, c = Q.init_state(config["batch"])
# h, c: 초기 장/단기 메모리 생성(전부 0)
estimateQs, _h, _c = Q(states, h, c) # (dBatch, dSeq, dAction)
# estimateQs, _h, _c: 현재 상태(states)에서의 Q 값 계산
estimateQs = estimateQs.gather(dim=-1, index=actions) # (dBatch, dSeq, 1)
# estimateQs: actions에 해당하는 Q를 가져옴

#################################################################################

print("estimateQ 형태:", estimateQs.shape)
print(estimateQs)
```

```python
# target Q 게산
## 장/단기 메모리만 추가되었고, 큰 틀은 DQN과 유사

################################### 코드 작성 ###################################

targetH, targetC = targetQ.init_state(config["batch"])
# targetH, targetC: 초기 장/단기 메모리 사용(모두 0으로 초기화)
nextTargetQs, _targetH, _targetC = targetQ(nextStates, targetH, targetC) # (dBatch, dSeq, dAction)
# nextTargetQs, _targetH, _targetC: 다음 상태(nextStates)의 Q 계산
nextTargetQs = nextTargetQs.max(dim=-1, keepdim=True).values # (dBatch, dSeq, 1)
# nextTargetQs: targetQ 네트워크로 다음 상태에 대한 Q값 중 가장 큰 값을 가져옴
## max_a' Q'(s', a')

targetQs = rewards + (1 - dones) * gamma * nextTargetQs # (dBatch, dSeq, 1)
# targetQs: 1-step TD target 방식으로 target Q 계산
## r + gamma * max_a' Q'(s', a')

#################################################################################

print("targetQ 형태:", targetQs.shape)
print("targetQ", targetQs)
```

```python
# MSE loss 계산 - DQN과 동일

################################### 코드 작성 ###################################

loss = F.mse_loss(targetQs.detach(), estimateQs)
# loss: targetQs와 estimateQs 간의 MSE loss 계산

#################################################################################

print(loss)
```

### 3-8. test score 계산

```python
# 성능 측정을 위해 한 episode 동안 reward를 합산해서 반환하는 함수 정의

def test(env, Q, dAction, epsilon=0.0):
    with torch.no_grad():
        score = 0.0
        state = env.reset()
        h, c = Q.init_state()
        while True:
            action, h, c = getAction(state, h, c, dAction, epsilon, Q)
            nextState, reward, done, _info = env.step(action)
            score += reward
            state = nextState
            if done:
                break
    return score

print("test score:", test(test_env, Q, dAction, config["epsilon_min"]))
# score 출력
```

### 3-9. DRQN의 전체 코드

```python
# 약 8분 소요

# 필요 라이브러리 로드
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import gym
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
# 지저분한 deprecation warning 문구 제거

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# 3-1. hyperparameter 정의
config = dict(
    env_name            = "Acrobot-v1",
    batch               = 8,
    buffer_size         = 100, # 버퍼의 최대 크기(episode 단위)
    lr                  = 0.001, # 학습률
    gamma               = 0.99, # discount(할인율)
    epsilon_init        = 0.1, # 초기 epsilon
    epsilon_min         = 0.001, # 최소 epsilon
    epsilon_decay       = 0.995,
    n_step              = 50000, # 학습 횟수
    n_train_start       = 8000,
    target_update_freq  = 100,  # target Q의 업데이트 주기(time step 단위)
    test_freq           = 100, # test 주기(time step 단위)
    max_sample_len      = 500, # sampling 시 최대 transition 시퀀스 길이
    sample_length       = 10, # sampling 시 시퀀스 길이
)

env      = gym.make(config["env_name"])
test_env = gym.make(config["env_name"])
dState = env.observation_space.shape[0] # 환경의 state_space 크기
dAction = env.action_space.n # 환경의 discrete action 개수

# 3-2. Q 네트워크 정의
class Qnet(nn.Module):
    def __init__(self, dState, dAction):
        super().__init__()
        self.dHidden = 128
        self.inputLayer = nn.Linear(in_features=dState, out_features=self.dHidden) # 입력층
        self.lstm = nn.LSTM(input_size=self.dHidden, hidden_size=self.dHidden, batch_first=True) # 은닉층=LSTM
        self.outputLayer = nn.Linear(in_features=self.dHidden, out_features=dAction) # 출력층

    def init_state(self, dBatch=1):
        h = torch.zeros([1, dBatch, self.dHidden]).to(DEVICE)
        c = torch.zeros([1, dBatch, self.dHidden]).to(DEVICE)
        # 초기 0으로 설정된 장/단기 메모리 반환
        return h, c

    def forward(self, x, h, c): # (dBatch, dSeq, dState), (1, dBatch, dHidden), (1, dBatch, dHidden)
        # 입력 데이터 x와 장/단기 메모리 h, c 를 받음
        x = F.relu(self.inputLayer(x)) # (dBatch, dSeq, dHidden)
        # 입력층 통과
        x, (newH, newC) = self.lstm(x, (h, c)) # (dBatch, dSeq, dHidden), (1, dBatch, dHidden), (1, dBatch, dHidden)
        # 은닉층=LSTM을 통과, 순환신경망이므로 내부적으로 sequnce 데이터가 순환하여 통과됨
        x = self.outputLayer(x) # (dBatch, dSeq, dAction)
        # 출력층 통과
        return x, newH, newC # (dBatch, dSeq, dState), (1, dBatch, dHidden), (1, dBatch, dHidden)
        # 출력값 및 장/단기 메모리를 함께 반환


Q = Qnet(dState=dState, dAction=dAction).to(DEVICE)
targetQ = Qnet(dState=dState, dAction=dAction).to(DEVICE) # target Q 정의
targetQ.load_state_dict(Q.state_dict())
optimizerQ = optim.Adam(Q.parameters(), lr=config["lr"])

# 3-3. episode buffer 정의
buffer = deque(maxlen=config["buffer_size"])

gamma = config["gamma"]
epsilon = config["epsilon_init"]

# 3-4. 행동 선택 함수 정의
def getAction(state, h, c, dAction, epsilon, Q): # (dState)
    with torch.no_grad():
        state = torch.from_numpy(state).float().to(DEVICE).view(1, 1, -1) # (1, 1, dState)
        q, newH, newC = Q(state, h, c) # (1, 1, dAction), (1, 1, dHidden), (1, 1, dHidden)
        if (random.random() > epsilon): # exploitation
            action = q.argmax(dim=-1).detach().cpu().item()
        else: # exploration
            action = np.random.randint(low=0, high=dAction) # (-)
    return action, newH, newC

# 3-8. test score 계산 함수 정의
def test(env, Q, dAction, epsilon=0.0):
    with torch.no_grad():
        score = 0.0
        state = env.reset()
        h, c = Q.init_state()
        while True:
            action, h, c = getAction(state, h, c, dAction, epsilon, Q)
            nextState, reward, done, _info = env.step(action)
            score += reward
            state = nextState
            if done:
                break
    return score

# 3-5. 학습 전, 샘플 수집
h, c = Q.init_state()
state = env.reset()
ep = []
for t in range(1, config["n_train_start"]+1):
    action, h, c = getAction(state, h, c, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    transition = (state, action, reward, nextState, done)
    ep.append(transition)
    state = nextState
    if done:
        state = env.reset()
        h, c = Q.init_state()
        buffer.append(ep)
        ep = []


# 메인 학습 루프
score = 0.0
scores3 = []
test_scores3 = []
losses3 = []
qs3 = []
# score 및 각종 성능 계산을 위한 변수 정의

h, c = Q.init_state()
state = env.reset() # (dState)
# 환경 초기화 및 초기 상태 획득
for t in range(1, config["n_step"]+1):
    # --- 환경 상호작용 ---
    action, h, c = getAction(state, h, c, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    transition = (state, action, reward, nextState, done)
    ep.append(transition)

    score += reward
    state = nextState
    if done: # episode 종료 시
        buffer.append(ep)
        state = env.reset()
        scores3.append(score)
        score = 0.0
        h, c = Q.init_state()
        ep = []
        # --- epsilon 업데이트 ---
        epsilon = max(config["epsilon_min"], epsilon*config["epsilon_decay"])

    # --- 3-6. batch 샘플링 ---
    episodes = random.sample(buffer, config["batch"])

    minStep = config["max_sample_len"] # 500
    for episode in episodes:
        minStep = min(minStep, len(episode))

    samples = []
    for episode in episodes:
        if minStep > config["sample_length"]:
            idx = np.random.randint(0, len(episode)-config["sample_length"]+1)
            sample = [np.array(item) for item in zip(*episode[idx:idx+config["sample_length"]])]
            samples.append(sample)
        else:
            idx = np.random.randint(0, len(episode)-minStep+1)
            sample = [np.array(item) for item in zip(*episode[idx:idx+minStep])]
            samples.append(sample)

    batch = []
    for item in zip(*samples):
        item = torch.from_numpy(np.stack(item)).float().to(DEVICE)
        batch.append(item)
    states, actions, rewards, nextStates, dones = batch
    actions = actions.view(config["batch"], -1, 1).long() # (dBatch, dSeq, 1)
    rewards = rewards.view(config["batch"], -1, 1) # (dBatch, dSeq, 1)
    dones = dones.view(config["batch"], -1, 1) # (dBatch, dSeq, 1)
    # 배치 샘플 준비

    # --- 3.7. loss 계산 ---
    # estimate
    qH, qC = Q.init_state(config["batch"])
    estimateQs, _qH, _qC = Q(states, qH, qC) # (dBatch, dSeq, dAction)
    estimateQs = estimateQs.gather(dim=-1, index=actions) # (dBatch, dSeq, 1)

    # target Q
    with torch.no_grad():
        targetqH, targetqC = targetQ.init_state(config["batch"])
        nextTargetQs, _targetqH, _targetqC = targetQ(nextStates, targetqH, targetqC) # (dBatch, dSeq, dAction)
        nextTargetQs = nextTargetQs.max(dim=-1, keepdim=True).values # (dBatch, dSeq, 1)
        # max_a Q'(s', a)
        targetQs = rewards + (1 - dones) * gamma * nextTargetQs # (dBatch, dSeq, 1)
        # r + gamma * max_a Q'(s', a)

    loss = F.mse_loss(targetQs.detach(), estimateQs)
    # MSE{ targetQ, Q }

    # --- 오차 역전파 및 매개변수 업데이트 ---
    optimizerQ.zero_grad()
    loss.backward()
    optimizerQ.step()
    losses3.append(loss.detach().cpu().item())
    qs3.append(estimateQs.detach().mean().cpu().item())

    # --- target Q 네트워크 업데이트 ---
    if t % config["target_update_freq"] == 0:
        targetQ.load_state_dict(Q.state_dict())
        # 정해진 주기마다 targetQ 를 Q 네트워크로 업데이트

    # --- test score 계산 ---
    if t % config["test_freq"] == 0:
        test_score = test(test_env, Q, dAction, epsilon=config["epsilon_min"])
        test_scores3.append(test_score)
        print(f'[time:{t}]test score: {test_score}, train loss: {losses3[-1]:.3f}, epsilon: {epsilon:.3f}')
        # 정해진 주기마다 현재 모델로 환경과 한 episode 상호작용후 score 계산
```

## 5.2 POMDP에서 DQN

```python
# 필요 라이브러리 로드

import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import gym
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
# 지저분한 deprecation warning 문구 제거

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# hyperparameter 정의
config = dict(
    env_name            = "CartPole-v1", # 환경
    batch               = 128,          # 배치 크기
    buffer_size         = 10000,        # 버퍼의 최대 크기(transition 단위)
    lr                  = 0.001,        # 학습률
    gamma               = 0.99,         # discount(할인율)
    epsilon_init        = 0.1,          # 초기 epsilon(=초기 탐험 비율)
    epsilon_min         = 0.001,        # 최소 epsilon(=최종 탐험 비율)
    epsilon_decay       = 0.995,        # epsilon 감소 비율(매 episode마다 epsilon에 곱해짐)
    n_step              = 50000,        # 학습 횟수 (time step 단위)
    n_train_start       = 8000,         # 초기 버퍼 채우기(time step 단위)
    target_update_freq  = 100,          # target Q의 업데이트 주기(time step 단위)
    test_freq           = 100,          # test 주기(time step 단위)
)

# 환경 및 테스트 환경 정의
env      = gym.make(config["env_name"])
test_env = gym.make(config["env_name"])
dState = env.observation_space.shape[0] - 2 # 환경의 state_space 크기
dAction = env.action_space.n # 환경의 discrete action 개수

# Q 네트워크 클래스 정의
class Qnet(nn.Module):
    def __init__(self, dState, dAction):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(in_features=dState, out_features=128),    # 입력층
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=128),       # 은닉층
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=dAction),   # 출력층
        )
        # 신경망 모델 정의 총 3개의 레이어로 구성
        # 입력(state 차원) -> 입력층(128) -> 은닉층(128) -> 출력층(action 차원)

    def forward(self, x): # 네트워크 처리 과정 정의
        return self.layers(x) # self.layers 신경망을 통과한 후 출력을 반환


# Q 및 target Q 네트워크 정의
Q = Qnet(dState=dState, dAction=dAction).to(DEVICE)
targetQ = Qnet(dState=dState, dAction=dAction).to(DEVICE) # target Q 정의
targetQ.load_state_dict(Q.state_dict())
optimizerQ = optim.Adam(Q.parameters(), lr=config["lr"])

# replay buffer 정의
replay_buffer = deque(maxlen=config["buffer_size"])

gamma = config["gamma"]
epsilon = config["epsilon_init"]

def getAction(state, dAction, epsilon, Q):
    with torch.no_grad(): # 본 함수는 역전파를 위한 gradient 계산이 필요 없으므로 gradient 계산을 제외시켜 연산량 줄임
        if (random.random() > epsilon): # exploitation(활용)
            state = torch.from_numpy(state).float().to(DEVICE)
            # Q 네트워크가 device에 존재하므로 입력인 state도 동일한 device로 이동시켜줌
            action = Q(state).argmax(dim=-1).detach().cpu().numpy()
            # state에서 모든 action에 대해 Q를 계산하고, 가장 높은 Q를 가진 action 선택
        else: # exploration(탐험)
            action = np.random.randint(low=0, high=dAction)
            # random하게 action 선택
    return action

def test(env, Q, dAction, epsilon):
    with torch.no_grad():
        score = 0.0
        state = env.reset()
        state = state[::2]
        while True:
            action = getAction(state, dAction, epsilon, Q)
            nextState, reward, done, _info = env.step(action)
            nextState = nextState[::2]
            score += reward
            state = nextState
            if done:
                break
    return score


# 학습 전, 샘플 수집
state = env.reset()
state = state[::2]
for t in range(1, config["n_train_start"]+1):
    action = getAction(state, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    nextState = nextState[::2]
    transition = (state, action, reward, nextState, done)
    replay_buffer.append(transition)
    state = nextState
    if done:
        state = env.reset()
        state = state[::2]

# 메인 학습 루프
score = 0.0
scores4 = []
test_scores4 = []
losses4 = []
qs4 = []
# score 및 각종 성능 계산을 위한 변수 정의

state = env.reset() # (dState)
state = state[::2]
# 환경 초기화 및 초기 상태 획득
for t in range(1, config["n_step"]+1):
    # --- 환경 상호작용 ---
    action = getAction(state, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    nextState = nextState[::2]
    transition = (state, action, reward, nextState, done)
    replay_buffer.append(transition)

    score += reward
    state = nextState
    if done: # episode 종료 시
        state = env.reset()
        state = state[::2]
        scores4.append(score)
        score = 0.0
        # --- epsilon 업데이트 ---
        epsilon = max(config["epsilon_min"], epsilon*config["epsilon_decay"])

    # --- batch 샘플링 ---
    transitions = random.sample(replay_buffer, config["batch"])
    batch = []
    for item in zip(*transitions):
        item = torch.from_numpy(np.stack(item)).float().to(DEVICE)
        batch.append(item)
    states, actions, rewards, nextStates, dones = batch # (dBatch, -)
    actions = actions.unsqueeze(dim=-1).long()
    rewards = rewards.unsqueeze(dim=-1)
    dones = dones.unsqueeze(dim=-1)
    # 배치 샘플 준비

    # --- loss 계산 ---
    # estimate
    estimateQs = Q(states).gather(dim=1, index=actions) # (dBatch, 1)
    # Q(s, a)

    # target
    with torch.no_grad():
        nextTargetQs = targetQ(nextStates).max(dim=-1, keepdim=True).values # (dBatch, 1)
        # max_a Q'(s', a)
        targetQs = rewards + (1 - dones) * gamma * nextTargetQs # (dBatch, 1)
        # r + gamma * max_a Q'(s', a)

    loss = F.mse_loss(targetQs.detach(), estimateQs) # scalar
    # MSE{ targetQ, Q }

    # --- 오차 역전파 및 매개변수 업데이트 ---
    optimizerQ.zero_grad()
    loss.backward()
    optimizerQ.step()
    losses4.append(loss.detach().cpu().item())
    qs4.append(estimateQs.detach().mean().cpu().item())

    # --- target Q 네트워크 업데이트 ---
    if t % config["target_update_freq"] == 0:
        targetQ.load_state_dict(Q.state_dict())
        # 정해진 주기마다 targetQ 를 Q 네트워크로 업데이트

    # --- test score 계산 ---
    if t % config["test_freq"] == 0:
        test_score = test(test_env, Q, dAction, epsilon=config["epsilon_min"])
        test_scores4.append(test_score)
        print(f'[time:{t}]test score: {test_score}, train loss: {losses4[-1]:.3f}, epsilon: {epsilon:.3f}')
        # 정해진 주기마다 현재 모델로 환경과 한 episode 상호작용후 score 계산
```

```python
# test score, loss, 평균 Q 를 그래프로 출력

y1 = np.array(test_scores4)
y2 = np.array(losses4)
y3 = np.array(qs4)
x1 = np.arange(len(y1))*config["test_freq"]
x2 = np.arange(len(y2))
x3 = np.arange(len(y3))

fig, ax = plt.subplots(1, 3, sharex=True, sharey=False, figsize=(15, 5))
fig.supxlabel("Time step")

ax[0].plot(x1, y1)
ax[0].set_ylabel("Score")
ax[0].grid()

ax[1].plot(x2, y2)
ax[1].set_ylabel("loss")
ax[1].grid()

ax[2].plot(x3, y3)
ax[2].set_ylabel("estimate Q(mean)")
ax[2].grid()

plt.tight_layout()
```

POMDP 환경에서는 DQN의 학습을 위한 충분한 정보를 제공하지 못하여 제대로 학습되지 않는 모습을 보여준다.

## 5.3 POMDP에서 DRQN

```python
# 필요 라이브러리 로드

import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import gym
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
# 지저분한 deprecation warning 문구 제거

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# hyperparameter 정의
config = dict(
    env_name            = "CartPole-v1",
    batch               = 8,
    buffer_size         = 100, # 버퍼의 최대 크기(episode 단위)
    lr                  = 0.001, # 학습률
    gamma               = 0.99, # discount(할인율)
    epsilon_init        = 0.1, # 초기 epsilon
    epsilon_min         = 0.001, # 최소 epsilon
    epsilon_decay       = 0.995,
    n_step              = 50000, # 학습 횟수
    n_train_start       = 8000,
    target_update_freq  = 100,  # target Q의 업데이트 주기(time step 단위)
    test_freq           = 100, # test 주기(time step 단위)
    max_sample_len      = 500, # sampling 시 최대 transition 시퀀스 길이
    sample_length       = 10, # sampling 시 시퀀스 길이
)

# 환경 및 테스트 환경 정의
env      = gym.make(config["env_name"])
test_env = gym.make(config["env_name"])
dState = env.observation_space.shape[0] - 2 # 환경의 state_space 크기
dAction = env.action_space.n # 환경의 discrete action 개수

# Q 네트워크 클래스 정의
class Qnet(nn.Module):
    def __init__(self, dState, dAction):
        super().__init__()
        self.dHidden = 128
        self.inputLayer = nn.Linear(in_features=dState, out_features=self.dHidden) # 입력층
        self.lstm = nn.LSTM(input_size=self.dHidden, hidden_size=self.dHidden, batch_first=True) # 은닉층=LSTM
        self.outputLayer = nn.Linear(in_features=self.dHidden, out_features=dAction) # 출력층

    def init_state(self, dBatch=1):
        h = torch.zeros([1, dBatch, self.dHidden]).to(DEVICE)
        c = torch.zeros([1, dBatch, self.dHidden]).to(DEVICE)
        # 초기 0으로 설정된 장/단기 메모리 반환
        return h, c

    def forward(self, x, h, c): # (dBatch, dSeq, dState), (1, dBatch, dHidden), (1, dBatch, dHidden)
        # 입력 데이터 x와 장/단기 메모리 h, c 를 받음
        x = F.relu(self.inputLayer(x)) # (dBatch, dSeq, dHidden)
        # 입력층 통과
        x, (newH, newC) = self.lstm(x, (h, c)) # (dBatch, dSeq, dHidden), (1, dBatch, dHidden), (1, dBatch, dHidden)
        # 은닉층=LSTM을 통과, 순환신경망이므로 내부적으로 sequnce 데이터가 순환하여 통과됨
        x = self.outputLayer(x) # (dBatch, dSeq, dAction)
        # 출력층 통과
        return x, newH, newC # (dBatch, dSeq, dState), (1, dBatch, dHidden), (1, dBatch, dHidden)
        # 출력값 및 장/단기 메모리를 함께 반환


# Q 및 target Q 네트워크 정의
Q = Qnet(dState=dState, dAction=dAction).to(DEVICE)
targetQ = Qnet(dState=dState, dAction=dAction).to(DEVICE) # target Q 정의
targetQ.load_state_dict(Q.state_dict())
optimizerQ = optim.Adam(Q.parameters(), lr=config["lr"])

# episode buffer 정의
buffer = deque(maxlen=config["buffer_size"])

gamma = config["gamma"]
epsilon = config["epsilon_init"]

def getAction(state, h, c, dAction, epsilon, Q): # (dState)
    with torch.no_grad():
        state = torch.from_numpy(state).float().to(DEVICE).view(1, 1, -1) # (1, 1, dState)
        q, newH, newC = Q(state, h, c) # (1, 1, dAction), (1, 1, dHidden), (1, 1, dHidden)
        if (random.random() > epsilon): # exploitation
            action = q.argmax(dim=-1).detach().cpu().item()
        else: # exploration
            action = np.random.randint(low=0, high=dAction) # (-)
    return action, newH, newC


def test(env, Q, dAction, epsilon=0.0):
    with torch.no_grad():
        score = 0.0
        state = env.reset()
        state = state[::2]
        h, c = Q.init_state()
        while True:
            action, h, c = getAction(state, h, c, dAction, epsilon, Q)
            nextState, reward, done, _info = env.step(action)
            nextState = nextState[::2]
            score += reward
            state = nextState
            if done:
                break
    return score


h, c = Q.init_state()
state = env.reset()
state = state[::2]
ep = []
for t in range(1, config["n_train_start"]+1):
    action, h, c = getAction(state, h, c, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    nextState = nextState[::2]
    transition = (state, action, reward, nextState, done)
    ep.append(transition)
    state = nextState
    if done:
        state = env.reset()
        state = state[::2]
        h, c = Q.init_state()
        buffer.append(ep)
        ep = []


# 메인 학습 루프
score = 0.0
scores5 = []
test_scores5 = []
losses5 = []
qs5 = []
# score 및 각종 성능 계산을 위한 변수 정의

h, c = Q.init_state()
state = env.reset() # (dState)
state = state[::2]
# 환경 초기화 및 초기 상태 획득
for t in range(1, config["n_step"]+1):
    # --- 환경 상호작용 ---
    action, h, c = getAction(state, h, c, dAction, epsilon, Q)
    nextState, reward, done, _info = env.step(action)
    nextState = nextState[::2]
    transition = (state, action, reward, nextState, done)
    ep.append(transition)

    score += reward
    state = nextState
    if done: # episode 종료 시
        buffer.append(ep)
        state = env.reset()
        state = state[::2]
        scores5.append(score)
        score = 0.0
        h, c = Q.init_state()
        ep = []
        # --- epsilon 업데이트 ---
        epsilon = max(config["epsilon_min"], epsilon*config["epsilon_decay"])

    # --- batch 샘플링 ---
    episodes = random.sample(buffer, config["batch"])

    minStep = config["max_sample_len"] # 500
    for episode in episodes:
        minStep = min(minStep, len(episode))

    samples = []
    for episode in episodes:
        if minStep > config["sample_length"]:
            idx = np.random.randint(0, len(episode)-config["sample_length"]+1)
            sample = [np.array(item) for item in zip(*episode[idx:idx+config["sample_length"]])]
            samples.append(sample)
        else:
            idx = np.random.randint(0, len(episode)-minStep+1)
            sample = [np.array(item) for item in zip(*episode[idx:idx+minStep])]
            samples.append(sample)

    batch = []
    for item in zip(*samples):
        item = torch.from_numpy(np.stack(item)).float().to(DEVICE)
        batch.append(item)
    states, actions, rewards, nextStates, dones = batch
    actions = actions.view(config["batch"], -1, 1).long() # (dBatch, dSeq, 1)
    rewards = rewards.view(config["batch"], -1, 1) # (dBatch, dSeq, 1)
    dones = dones.view(config["batch"], -1, 1) # (dBatch, dSeq, 1)
    # 배치 샘플 준비

    # --- loss 계산 ---
    # estimate
    qH, qC = Q.init_state(config["batch"])
    estimateQs, _qH, _qC = Q(states, qH, qC) # (dBatch, dSeq, dAction)
    estimateQs = estimateQs.gather(dim=-1, index=actions) # (dBatch, dSeq, 1)

    # target Q
    with torch.no_grad():
        targetqH, targetqC = targetQ.init_state(config["batch"])
        nextTargetQs, _targetqH, _targetqC = targetQ(nextStates, targetqH, targetqC) # (dBatch, dSeq, dAction)
        nextTargetQs = nextTargetQs.max(dim=-1, keepdim=True).values # (dBatch, dSeq, 1)
        # max_a Q'(s', a)
        targetQs = rewards + (1 - dones) * gamma * nextTargetQs # (dBatch, dSeq, 1)
        # r + gamma * max_a Q'(s', a)

    loss = F.mse_loss(targetQs.detach(), estimateQs)
    # MSE{ targetQ, Q }

    # --- 오차 역전파 및 매개변수 업데이트 ---
    optimizerQ.zero_grad()
    loss.backward()
    optimizerQ.step()
    losses5.append(loss.detach().cpu().item())
    qs5.append(estimateQs.detach().mean().cpu().item())

    # --- target Q 네트워크 업데이트 ---
    if t % config["target_update_freq"] == 0:
        targetQ.load_state_dict(Q.state_dict())
        # 정해진 주기마다 targetQ 를 Q 네트워크로 업데이트

    # --- test score 계산 ---
    if t % config["test_freq"] == 0:
        test_score = test(test_env, Q, dAction, epsilon=config["epsilon_min"])
        test_scores5.append(test_score)
        print(f'[time:{t}]test score: {test_score}, train loss: {losses5[-1]:.3f}, epsilon: {epsilon:.3f}')
        # 정해진 주기마다 현재 모델로 환경과 한 episode 상호작용후 score 계산
```

```python
y1 = np.array(test_scores5)
y2 = np.array(losses5)
y3 = np.array(qs5)
x1 = np.arange(len(y1))*config["test_freq"]
x2 = np.arange(len(y2))
x3 = np.arange(len(y3))

fig, ax = plt.subplots(1, 3, sharex=True, sharey=False, figsize=(15, 5))
fig.supxlabel("Time step")

ax[0].plot(x1, y1)
ax[0].set_ylabel("Score")
ax[0].grid()

ax[1].plot(x2, y2)
ax[1].set_ylabel("loss")
ax[1].grid()

ax[2].plot(x3, y3)
ax[2].set_ylabel("estimate Q(mean)")
ax[2].grid()

plt.tight_layout()
```
