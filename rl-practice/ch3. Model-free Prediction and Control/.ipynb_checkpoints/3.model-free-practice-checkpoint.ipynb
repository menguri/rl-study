```python
import matplotlib.animation as animation
import ipywidgets as widgets
from IPython.display import display

def simulate_policy(env, Q, fps=5):
    """
    학습된 Q 테이블을 사용하여 정책을 실행하고, 
    Matplotlib 애니메이션을 활용하여 자연스럽게 시뮬레이션을 출력하는 함수.

    Args:
        env: OpenAI Gym 환경
        Q: 학습된 Q 테이블 (numpy array)
        fps: 초당 프레임 수 (기본값: 5, 낮을수록 느리게 재생)
    """

    def start_simulation(_):
        """버튼 클릭 시 시뮬레이션 실행"""
        state = env.reset()[0]  # 환경 초기화
        frames = []  # 프레임 저장 리스트
        total_reward = 0  # 총 보상

        while True:
            action = np.argmax(Q[state, :])  # 최적 행동 선택
            new_state, reward, done, truncated, info = env.step(action)

            # 현재 상태를 프레임으로 저장
            frame = env.render()
            frames.append(frame)

            total_reward += reward
            state = new_state

            if done:
                print(f"🚀 에피소드 종료! 총 보상: {total_reward}\n")
                break

        # 📌 Matplotlib 애니메이션 설정 (fps 조절 가능)
        fig, ax = plt.subplots(figsize=(6, 6))
        im = ax.imshow(frames[0])  # 첫 번째 프레임 설정
        ax.axis("off")  # 축 숨기기

        def update(frame):
            """이미지를 업데이트하는 함수"""
            im.set_array(frame)
            return im

        interval = 1000 / fps  # 초당 프레임 수(fps)에 맞춰 속도 조절
        ani = animation.FuncAnimation(fig, update, frames=frames, interval=interval, blit=False)
        plt.show()

    # 🟢 Jupyter Notebook에서 버튼 추가
    button = widgets.Button(description="Start Simulation", button_style="success")
    button.on_click(start_simulation)
    display(button)
```

```python
def make_plot_list(total_reward):
  window = []
  for i in range(num_episodes):                   #총 에피소드의 수 만큼 반복
    if i <= 50:                                   #만약 50에피소드 이전 결과라면, 그동안의 평균 값을 계산
      window.append(np.mean(total_reward[:i]))
    else:                                         #50에피소드 이후 결과들은 이전 50에피소드의 평균으로 계산
      window.append(np.mean(total_reward[i-50:i]))
  return window
```

# HW1. e-greedy 방법으로 Q-learning 작성

# Q-learning의 식은 다음과 같음
image.png


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # seaborn을 사용하여 히트맵을 작성
from matplotlib.colors import ListedColormap
import gym

# CliffWalking 환경 생성

env = gym.make("CliffWalking-v0")

# Q table을 observation과 action space의 크기로 초기화
Q = np.zeros((env.observation_space.n, env.action_space.n))

#하이퍼파라미터는 아래와 같이 고정
discount_factor = 0.99
num_episodes = 2000
alpha = 0.9

total_reward = []
total_reward_sum = []
for i in range(num_episodes):

    # 환경을 초기화
    state, _ = env.reset()

    # reward 의 합을 0으로 초기화
    reward_sum = 0

    # 에피소드가 끝났음을 알리는 done_mask를 False로 초기화
    done = False

    #학습이 진행될수록 점점 줄어드는 epsilon 값을 정의
    #이 때, epsilon 값은 50 episode마다 감소하도록 설정
    e = 1. / ((i//50) + 1)

    while not done:
        #epsilon의 확률로 무작위 행동을 선택
        if np.random.rand(1) < e:
            action = env.action_space.sample()
        #이외의 경우엔 Q값을 따라가도록 행동을 선택
        else:
            action = np.argmax(Q[state, :])

        #선택된 행동을 환경에서 실행
        new_state, reward, done, truncated, info = env.step(action)


        #Q 값 업데이트.(위에 있는 식과 동일하게 작성)
        Q[state, action] = Q[state, action] + alpha * (reward + discount_factor * np.max(Q[new_state, :]) - Q[state, action])

        reward_sum += reward
        state = new_state

    total_reward.append(reward_sum)
    total_reward_sum.append(sum(total_reward)/(i+1))
```

```python
print("Final Q-Table Values")
window = make_plot_list(total_reward)
plt.figure()
plt.plot(window)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.show()

env = gym.make("CliffWalking-v0", render_mode="rgb_array")
for i in range(3):
    simulate_policy(env, Q)

print("Simulation complete.")


env.close()
```

```python
# Sarsa with Noise
def cliff_sarsa(env, gamma, num_episodes, alpha):

    Q = np.zeros([env.observation_space.n, env.action_space.n])

    # ✅ 실행 시간 측정 시작
    execution_time = 0.0
    start_time = time.time()

    
    total_reward = []
    total_reward_sum = []

    best_score = -13  # 현재까지의 최고 점수
    best_episode = -1  # 최고 점수를 달성한 에피소드
    total_computations = 0  # 총 Q 업데이트 횟수
    
    for i in tqdm(range(num_episodes)):
    
        # 환경 초기화
        state, _ = env.reset()
    
        # reward 의 합을 0으로 초기화
        reward_sum = 0
    
        # 에피소드가 끝났음을 알리는 done_mask를 False로 초기화
        done = False
    
        # 랜덤 노이즈 작성.처음엔 1의 값에서, 에피소드 10마다 값이 점차 줄어들도록 작성
        decay_ratio = 1./((i//10) + 1) 
        random_noise = np.random.randn(1, env.action_space.n) * decay_ratio

        #이 때, epsilon 값은 50 episode마다 감소하도록 설정
        e = 1. / ((i//50) + 1)
    
        action = np.argmax(Q[state, :] + random_noise)
    
        while not done:
            new_state, reward, done, truncated, info = env.step(action)
    
            #Q값에 랜덤 노이즈를 추가한 값들 중, 가장 큰 값을 선택하도록 작성
            #epsilon의 확률로 무작위 행동을 선택
            if np.random.rand(1) < e:
                new_action = env.action_space.sample()
            #이외의 경우엔 Q값을 따라가도록 행동을 선택
            else:
                new_action = np.argmax(Q[new_state, :] + random_noise)
    
            #SARSA 업데이트 방법으로 Q 테이블 업데이트 코드를 작성
            Q[state, action] = Q[state, action] + alpha * (reward + gamma * Q[new_state, new_action] - Q[state, action])
    
            reward_sum += reward
            state = new_state
            action = new_action
    
        total_reward.append(reward_sum)
        total_reward_sum.append(sum(total_reward)/(i+1))

        # ✅ Best Score 기록
        if reward_sum > best_score:
            best_score = reward_sum
            best_episode = i + 1  # 0-index가 아니라 1-index로 기록
            execution_time = time.time() - start_time
            break

    return Q, total_reward, total_reward_sum, execution_time, best_score, best_episode
```

```python
# Sarsa-lambda with Noise
def cliff_sarsa_lambda(env, gamma, num_episodes, alpha, lam=0.8):

    Q = np.zeros([env.observation_space.n, env.action_space.n])

    # ✅ 실행 시간 측정 시작
    execution_time = 0.0
    start_time = time.time()

    
    total_reward = []
    total_reward_sum = []

    best_score = -13  # 현재까지의 최고 점수
    best_episode = -1  # 최고 점수를 달성한 에피소드
    total_computations = 0  # 총 Q 업데이트 횟수

    for i in tqdm(range(num_episodes)):

        # episode
        episode = []
    
        # 환경 초기화
        state, _ = env.reset()
    
        # reward 의 합을 0으로 초기화
        reward_sum = 0
    
        # 에피소드가 끝났음을 알리는 done_mask를 False로 초기화
        done = False
    
        # 랜덤 노이즈 작성.처음엔 1의 값에서, 에피소드 10마다 값이 점차 줄어들도록 작성
        decay_ratio = 1./((i//10) + 1) 
        random_noise = np.random.randn(1, env.action_space.n) * decay_ratio

        #이 때, epsilon 값은 50 episode마다 감소하도록 설정
        e = 1. / ((i//50) + 1)
    
        action = np.argmax(Q[state, :] + random_noise)

        eligibility_traces = np.zeros_like(Q)
        
        while not done:
            new_state, reward, done, truncated, info = env.step(action)
    
            #Q값에 랜덤 노이즈를 추가한 값들 중, 가장 큰 값을 선택하도록 작성
            #epsilon의 확률로 무작위 행동을 선택
            if np.random.rand(1) < e:
                new_action = env.action_space.sample()
            #이외의 경우엔 Q값을 따라가도록 행동을 선택
            else:
                new_action = np.argmax(Q[new_state, :] + random_noise)
    
            # ✅ TD Error 계산
            next_value = 0 if done else Q[new_state, new_action]
            td_error = reward + gamma * next_value - Q[state, action]

            # ✅ Eligibility Traces 업데이트
            eligibility_traces[state, action] += 1  # 방문한 상태-행동 쌍만 증가

            # ✅ Q 테이블 업데이트 (방문한 상태-행동만 업데이트)
            Q += alpha * td_error * eligibility_traces

            # ✅ Eligibility Traces Decay (λ 적용)
            eligibility_traces *= gamma * lam

            reward_sum += reward
            state = new_state
            action = new_action

        total_reward.append(reward_sum)
        total_reward_sum.append(np.mean(total_reward))

        # ✅ Best Score 기록
        if reward_sum > best_score:
            best_score = reward_sum
            best_episode = i + 1  # 0-index가 아니라 1-index로 기록
            execution_time = time.time() - start_time
            break
    
    return Q, total_reward, total_reward_sum, execution_time, best_score, best_episode
```

```python
# Q-learning with Noise
def cliff_qlearning(env, gamma, num_episodes, alpha):

    Q = np.zeros([env.observation_space.n, env.action_space.n])

    # ✅ 실행 시간 측정 시작
    execution_time = 0.0
    start_time = time.time()

    
    total_reward = []
    total_reward_sum = []

    best_score = -13  # 현재까지의 최고 점수
    best_episode = -1  # 최고 점수를 달성한 에피소드
    total_computations = 0  # 총 Q 업데이트 횟수
    
    for i in tqdm(range(num_episodes)):
    
        # 환경 초기화
        state, _ = env.reset()
    
        # reward 의 합을 0으로 초기화
        reward_sum = 0
    
        # 에피소드가 끝났음을 알리는 done_mask를 False로 초기화
        done = False
    
        # 랜덤 노이즈 작성.처음엔 1의 값에서, 에피소드 10마다 값이 점차 줄어들도록 작성
        decay_ratio = 1./((i//10) + 1) 
        random_noise = np.random.randn(1, env.action_space.n) * decay_ratio

        #이 때, epsilon 값은 50 episode마다 감소하도록 설정
        e = 1. / ((i//50) + 1)
    
        while not done:

            #epsilon의 확률로 무작위 행동을 선택
            if np.random.rand(1) < e:
                action = env.action_space.sample()
            #이외의 경우엔 Q값을 따라가도록 행동을 선택
            else:
                action = np.argmax(Q[state, :] + random_noise)
                
            new_state, reward, done, truncated, info = env.step(action)
    
            #Q값에 랜덤 노이즈를 추가한 값들 중, 가장 큰 값을 선택하도록 작성
            new_action = np.argmax(Q[new_state, :] + random_noise)
    
            #SARSA 업데이트 방법으로 Q 테이블 업데이트 코드를 작성
            Q[state, action] = Q[state, action] + alpha * (reward + gamma * np.max(Q[new_state, :]) - Q[state, action])
    
            reward_sum += reward
            state = new_state
    
        total_reward.append(reward_sum)
        total_reward_sum.append(sum(total_reward)/(i+1))

        # ✅ Best Score 기록
        if reward_sum > best_score:
            best_score = reward_sum
            best_episode = i + 1  # 0-index가 아니라 1-index로 기록
            execution_time = time.time() - start_time
            break
    
    return Q, total_reward, total_reward_sum, execution_time, best_score, best_episode
```

```python
from gym import spaces

class CustomCliffWalkingEnv(gym.Env):
    """
    CliffWalking-v0을 기반으로 한 커스텀 환경.
    - 기존 클리프에 추가로 "구멍(Holes)"을 중간에 배치.
    - 구멍에 도달하면 시작점으로 되돌아가고 보상을 새롭게 설정.
    """

    def __init__(self):
        super(CustomCliffWalkingEnv, self).__init__()

        self.shape = (4, 12)  # 기존 크기 유지
        self.start_state = (3, 0)  # 시작 위치
        self.goal_state = (3, 11)  # 목표 위치
        self.cliff = [(3, i) for i in range(1, 11)]  # 기존 Cliff
        self.holes = [(0, 8), (1, 10)]  # ✅ 추가된 "구멍(Holes)" 좌표 (여기 도달하면 리셋)
        
        self.observation_space = spaces.Discrete(self.shape[0] * self.shape[1])
        self.action_space = spaces.Discrete(4)  # 0: 위, 1: 아래, 2: 오른쪽, 3: 왼쪽

        self.reset()

    def reset(self):
        self.state = self.start_state  # 시작점으로 리셋
        return self.get_state_index(), {}

    def step(self, action):
        row, col = self.state

        # ✅ 행동 정의 (Up, Down, Right, Left)
        if action == 0 and row > 0:  # 위로 이동
            row -= 1
        elif action == 1 and row < self.shape[0] - 1:  # 아래로 이동
            row += 1
        elif action == 2 and col < self.shape[1] - 1:  # 오른쪽 이동
            col += 1
        elif action == 3 and col > 0:  # 왼쪽 이동
            col -= 1

        self.state = (row, col)

        # ✅ 보상 & 종료 조건 설정
        if self.state == self.goal_state:
            reward = 100  # 목표 도착
            done = True
        elif self.state in self.cliff:
            reward = -100  # Cliff에 떨어짐
            self.state = self.start_state  # 시작점으로 리셋
            done = False
        elif self.state in self.holes:
            reward = -50  # 구멍에 빠짐
            self.state = self.start_state  # 시작점으로 리셋
            done = False
        else:
            reward = -2  # 일반 이동 (더 어렵게 설정)
            done = False

        return self.get_state_index(), reward, done, False, {}

    def get_state_index(self):
        """2D 좌표를 1D 상태 인덱스로 변환"""
        return self.state[0] * self.shape[1] + self.state[1]
```

```python
# 새로운 환경 생성
env = CustomCliffWalkingEnv()
discount_factor = 0.99
num_episodes = 300
alpha = 0.9
```
