```python
import matplotlib.animation as animation
import ipywidgets as widgets
from IPython.display import display

def simulate_policy(env, Q, fps=5):
    """
    ÌïôÏäµÎêú Q ÌÖåÏù¥Î∏îÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Ï†ïÏ±ÖÏùÑ Ïã§ÌñâÌïòÍ≥†, 
    Matplotlib Ïï†ÎãàÎ©îÏù¥ÏÖòÏùÑ ÌôúÏö©ÌïòÏó¨ ÏûêÏó∞Ïä§ÎüΩÍ≤å ÏãúÎÆ¨Î†àÏù¥ÏÖòÏùÑ Ï∂úÎ†•ÌïòÎäî Ìï®Ïàò.

    Args:
        env: OpenAI Gym ÌôòÍ≤Ω
        Q: ÌïôÏäµÎêú Q ÌÖåÏù¥Î∏î (numpy array)
        fps: Ï¥àÎãπ ÌîÑÎ†àÏûÑ Ïàò (Í∏∞Î≥∏Í∞í: 5, ÎÇÆÏùÑÏàòÎ°ù ÎäêÎ¶¨Í≤å Ïû¨ÏÉù)
    """

    def start_simulation(_):
        """Î≤ÑÌäº ÌÅ¥Î¶≠ Ïãú ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ"""
        state = env.reset()[0]  # ÌôòÍ≤Ω Ï¥àÍ∏∞Ìôî
        frames = []  # ÌîÑÎ†àÏûÑ Ï†ÄÏû• Î¶¨Ïä§Ìä∏
        total_reward = 0  # Ï¥ù Î≥¥ÏÉÅ

        while True:
            action = np.argmax(Q[state, :])  # ÏµúÏ†Å ÌñâÎèô ÏÑ†ÌÉù
            new_state, reward, done, truncated, info = env.step(action)

            # ÌòÑÏû¨ ÏÉÅÌÉúÎ•º ÌîÑÎ†àÏûÑÏúºÎ°ú Ï†ÄÏû•
            frame = env.render()
            frames.append(frame)

            total_reward += reward
            state = new_state

            if done:
                print(f"üöÄ ÏóêÌîºÏÜåÎìú Ï¢ÖÎ£å! Ï¥ù Î≥¥ÏÉÅ: {total_reward}\n")
                break

        # üìå Matplotlib Ïï†ÎãàÎ©îÏù¥ÏÖò ÏÑ§Ï†ï (fps Ï°∞Ï†à Í∞ÄÎä•)
        fig, ax = plt.subplots(figsize=(6, 6))
        im = ax.imshow(frames[0])  # Ï≤´ Î≤àÏß∏ ÌîÑÎ†àÏûÑ ÏÑ§Ï†ï
        ax.axis("off")  # Ï∂ï Ïà®Í∏∞Í∏∞

        def update(frame):
            """Ïù¥ÎØ∏ÏßÄÎ•º ÏóÖÎç∞Ïù¥Ìä∏ÌïòÎäî Ìï®Ïàò"""
            im.set_array(frame)
            return im

        interval = 1000 / fps  # Ï¥àÎãπ ÌîÑÎ†àÏûÑ Ïàò(fps)Ïóê ÎßûÏ∂∞ ÏÜçÎèÑ Ï°∞Ï†à
        ani = animation.FuncAnimation(fig, update, frames=frames, interval=interval, blit=False)
        plt.show()

    # üü¢ Jupyter NotebookÏóêÏÑú Î≤ÑÌäº Ï∂îÍ∞Ä
    button = widgets.Button(description="Start Simulation", button_style="success")
    button.on_click(start_simulation)
    display(button)
```

```python
def make_plot_list(total_reward):
  window = []
  for i in range(num_episodes):                   #Ï¥ù ÏóêÌîºÏÜåÎìúÏùò Ïàò ÎßåÌÅº Î∞òÎ≥µ
    if i <= 50:                                   #ÎßåÏïΩ 50ÏóêÌîºÏÜåÎìú Ïù¥Ï†Ñ Í≤∞Í≥ºÎùºÎ©¥, Í∑∏ÎèôÏïàÏùò ÌèâÍ∑† Í∞íÏùÑ Í≥ÑÏÇ∞
      window.append(np.mean(total_reward[:i]))
    else:                                         #50ÏóêÌîºÏÜåÎìú Ïù¥ÌõÑ Í≤∞Í≥ºÎì§ÏùÄ Ïù¥Ï†Ñ 50ÏóêÌîºÏÜåÎìúÏùò ÌèâÍ∑†ÏúºÎ°ú Í≥ÑÏÇ∞
      window.append(np.mean(total_reward[i-50:i]))
  return window
```

# HW1. e-greedy Î∞©Î≤ïÏúºÎ°ú Q-learning ÏûëÏÑ±

# Q-learningÏùò ÏãùÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùå

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # seabornÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÌûàÌä∏ÎßµÏùÑ ÏûëÏÑ±
from matplotlib.colors import ListedColormap
import gym

# CliffWalking ÌôòÍ≤Ω ÏÉùÏÑ±

env = gym.make("CliffWalking-v0")

# Q tableÏùÑ observationÍ≥º action spaceÏùò ÌÅ¨Í∏∞Î°ú Ï¥àÍ∏∞Ìôî
Q = np.zeros((env.observation_space.n, env.action_space.n))

#ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Îäî ÏïÑÎûòÏôÄ Í∞ôÏù¥ Í≥†Ï†ï
discount_factor = 0.99
num_episodes = 2000
alpha = 0.9

total_reward = []
total_reward_sum = []
for i in range(num_episodes):

    # ÌôòÍ≤ΩÏùÑ Ï¥àÍ∏∞Ìôî
    state, _ = env.reset()

    # reward Ïùò Ìï©ÏùÑ 0ÏúºÎ°ú Ï¥àÍ∏∞Ìôî
    reward_sum = 0

    # ÏóêÌîºÏÜåÎìúÍ∞Ä ÎÅùÎÇ¨ÏùåÏùÑ ÏïåÎ¶¨Îäî done_maskÎ•º FalseÎ°ú Ï¥àÍ∏∞Ìôî
    done = False

    #ÌïôÏäµÏù¥ ÏßÑÌñâÎê†ÏàòÎ°ù Ï†êÏ†ê Ï§ÑÏñ¥ÎìúÎäî epsilon Í∞íÏùÑ Ï†ïÏùò
    #Ïù¥ Îïå, epsilon Í∞íÏùÄ 50 episodeÎßàÎã§ Í∞êÏÜåÌïòÎèÑÎ°ù ÏÑ§Ï†ï
    e = 1. / ((i//50) + 1)

    while not done:
        #epsilonÏùò ÌôïÎ•†Î°ú Î¨¥ÏûëÏúÑ ÌñâÎèôÏùÑ ÏÑ†ÌÉù
        if np.random.rand(1) < e:
            action = env.action_space.sample()
        #Ïù¥Ïô∏Ïùò Í≤ΩÏö∞Ïóî QÍ∞íÏùÑ Îî∞ÎùºÍ∞ÄÎèÑÎ°ù ÌñâÎèôÏùÑ ÏÑ†ÌÉù
        else:
            action = np.argmax(Q[state, :])

        #ÏÑ†ÌÉùÎêú ÌñâÎèôÏùÑ ÌôòÍ≤ΩÏóêÏÑú Ïã§Ìñâ
        new_state, reward, done, truncated, info = env.step(action)


        #Q Í∞í ÏóÖÎç∞Ïù¥Ìä∏.(ÏúÑÏóê ÏûàÎäî ÏãùÍ≥º ÎèôÏùºÌïòÍ≤å ÏûëÏÑ±)
        Q[state, action] = Q[state, action] + alpha * (reward + discount_factor * np.max(Q[new_state, :]) - Q[state, action])

        reward_sum += reward
        state = new_state

    total_reward.append(reward_sum)
    total_reward_sum.append(sum(total_reward)/(i+1))
```

```python
print("Final Q-Table Values")
window = make_plot_list(total_reward)
plt.figure()
plt.plot(window)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.show()

env = gym.make("CliffWalking-v0", render_mode="rgb_array")
for i in range(3):
    simulate_policy(env, Q)

print("Simulation complete.")


env.close()
```

# HW2. random noiseÏùò SARSA ÏßÑÌñâ

```python
# CliffWalking ÌôòÍ≤Ω ÏÉùÏÑ±
env = gym.make("CliffWalking-v0")

# Q tableÏùÑ observationÍ≥º action spaceÏùò ÌÅ¨Í∏∞Î°ú Ï¥àÍ∏∞Ìôî
Q = np.zeros([env.observation_space.n, env.action_space.n])

#ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Îäî ÏïÑÎûòÏôÄ Í∞ôÏù¥ ÏÇ¨Ïö©
discount_factor = 0.99
num_episodes = 2000
alpha = 0.9

total_reward = []
total_reward_sum = []
for i in range(num_episodes):

    # ÌôòÍ≤Ω Ï¥àÍ∏∞Ìôî
    state, _ = env.reset()

    # reward Ïùò Ìï©ÏùÑ 0ÏúºÎ°ú Ï¥àÍ∏∞Ìôî
    reward_sum = 0

    # ÏóêÌîºÏÜåÎìúÍ∞Ä ÎÅùÎÇ¨ÏùåÏùÑ ÏïåÎ¶¨Îäî done_maskÎ•º FalseÎ°ú Ï¥àÍ∏∞Ìôî
    done = False

    # ÎûúÎç§ ÎÖ∏Ïù¥Ï¶à ÏûëÏÑ±.Ï≤òÏùåÏóî 1Ïùò Í∞íÏóêÏÑú, ÏóêÌîºÏÜåÎìú 10ÎßàÎã§ Í∞íÏù¥ Ï†êÏ∞® Ï§ÑÏñ¥Îì§ÎèÑÎ°ù ÏûëÏÑ±
    decay_ratio = 1./((i//10) + 1) 
    random_noise = np.random.randn(1, env.action_space.n) * decay_ratio

    action = np.argmax(Q[state, :] + random_noise)

    while not done:
        new_state, reward, done, truncated, info = env.step(action)

        #QÍ∞íÏóê ÎûúÎç§ ÎÖ∏Ïù¥Ï¶àÎ•º Ï∂îÍ∞ÄÌïú Í∞íÎì§ Ï§ë, Í∞ÄÏû• ÌÅ∞ Í∞íÏùÑ ÏÑ†ÌÉùÌïòÎèÑÎ°ù ÏûëÏÑ±
        new_action = np.argmax(Q[new_state, :] + random_noise)

        #SARSA ÏóÖÎç∞Ïù¥Ìä∏ Î∞©Î≤ïÏúºÎ°ú Q ÌÖåÏù¥Î∏î ÏóÖÎç∞Ïù¥Ìä∏ ÏΩîÎìúÎ•º ÏûëÏÑ±
        Q[state, action] = Q[state, action] + alpha * (reward + discount_factor * Q[new_state, new_action] - Q[state, action])

        reward_sum += reward
        state = new_state
        action = new_action

    total_reward.append(reward_sum)
    total_reward_sum.append(sum(total_reward)/(i+1))

```

```python
print("Final Q-Table Values")


window = make_plot_list(total_reward)
plt.figure()
plt.plot(window)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.show()

env = gym.make("CliffWalking-v0", render_mode="rgb_array")
for i in range(3):
    simulate_policy(env, Q)

print("Simulation complete.")


env.close()
```

```python
# Sarsa with Noise
def cliff_sarsa(env, gamma, num_episodes, alpha):

    Q = np.zeros([env.observation_space.n, env.action_space.n])

    # ‚úÖ Ïã§Ìñâ ÏãúÍ∞Ñ Ï∏°Ï†ï ÏãúÏûë
    execution_time = 0.0
    start_time = time.time()

    
    total_reward = []
    total_reward_sum = []

    best_score = -13  # ÌòÑÏû¨ÍπåÏßÄÏùò ÏµúÍ≥† Ï†êÏàò
    best_episode = -1  # ÏµúÍ≥† Ï†êÏàòÎ•º Îã¨ÏÑ±Ìïú ÏóêÌîºÏÜåÎìú
    total_computations = 0  # Ï¥ù Q ÏóÖÎç∞Ïù¥Ìä∏ ÌöüÏàò
    
    for i in tqdm(range(num_episodes)):
    
        # ÌôòÍ≤Ω Ï¥àÍ∏∞Ìôî
        state, _ = env.reset()
    
        # reward Ïùò Ìï©ÏùÑ 0ÏúºÎ°ú Ï¥àÍ∏∞Ìôî
        reward_sum = 0
    
        # ÏóêÌîºÏÜåÎìúÍ∞Ä ÎÅùÎÇ¨ÏùåÏùÑ ÏïåÎ¶¨Îäî done_maskÎ•º FalseÎ°ú Ï¥àÍ∏∞Ìôî
        done = False
    
        # ÎûúÎç§ ÎÖ∏Ïù¥Ï¶à ÏûëÏÑ±.Ï≤òÏùåÏóî 1Ïùò Í∞íÏóêÏÑú, ÏóêÌîºÏÜåÎìú 10ÎßàÎã§ Í∞íÏù¥ Ï†êÏ∞® Ï§ÑÏñ¥Îì§ÎèÑÎ°ù ÏûëÏÑ±
        decay_ratio = 1./((i//10) + 1) 
        random_noise = np.random.randn(1, env.action_space.n) * decay_ratio

        #Ïù¥ Îïå, epsilon Í∞íÏùÄ 50 episodeÎßàÎã§ Í∞êÏÜåÌïòÎèÑÎ°ù ÏÑ§Ï†ï
        e = 1. / ((i//50) + 1)
    
        action = np.argmax(Q[state, :] + random_noise)
    
        while not done:
            new_state, reward, done, truncated, info = env.step(action)
    
            #QÍ∞íÏóê ÎûúÎç§ ÎÖ∏Ïù¥Ï¶àÎ•º Ï∂îÍ∞ÄÌïú Í∞íÎì§ Ï§ë, Í∞ÄÏû• ÌÅ∞ Í∞íÏùÑ ÏÑ†ÌÉùÌïòÎèÑÎ°ù ÏûëÏÑ±
            #epsilonÏùò ÌôïÎ•†Î°ú Î¨¥ÏûëÏúÑ ÌñâÎèôÏùÑ ÏÑ†ÌÉù
            if np.random.rand(1) < e:
                new_action = env.action_space.sample()
            #Ïù¥Ïô∏Ïùò Í≤ΩÏö∞Ïóî QÍ∞íÏùÑ Îî∞ÎùºÍ∞ÄÎèÑÎ°ù ÌñâÎèôÏùÑ ÏÑ†ÌÉù
            else:
                new_action = np.argmax(Q[new_state, :] + random_noise)
    
            #SARSA ÏóÖÎç∞Ïù¥Ìä∏ Î∞©Î≤ïÏúºÎ°ú Q ÌÖåÏù¥Î∏î ÏóÖÎç∞Ïù¥Ìä∏ ÏΩîÎìúÎ•º ÏûëÏÑ±
            Q[state, action] = Q[state, action] + alpha * (reward + gamma * Q[new_state, new_action] - Q[state, action])
    
            reward_sum += reward
            state = new_state
            action = new_action
    
        total_reward.append(reward_sum)
        total_reward_sum.append(sum(total_reward)/(i+1))

        # ‚úÖ Best Score Í∏∞Î°ù
        if reward_sum > best_score:
            best_score = reward_sum
            best_episode = i + 1  # 0-indexÍ∞Ä ÏïÑÎãàÎùº 1-indexÎ°ú Í∏∞Î°ù
            execution_time = time.time() - start_time
            break

    return Q, total_reward, total_reward_sum, execution_time, best_score, best_episode
```

```python
# Sarsa-lambda with Noise
def cliff_sarsa_lambda(env, gamma, num_episodes, alpha, lam=0.8):

    Q = np.zeros([env.observation_space.n, env.action_space.n])

    # ‚úÖ Ïã§Ìñâ ÏãúÍ∞Ñ Ï∏°Ï†ï ÏãúÏûë
    execution_time = 0.0
    start_time = time.time()

    
    total_reward = []
    total_reward_sum = []

    best_score = -13  # ÌòÑÏû¨ÍπåÏßÄÏùò ÏµúÍ≥† Ï†êÏàò
    best_episode = -1  # ÏµúÍ≥† Ï†êÏàòÎ•º Îã¨ÏÑ±Ìïú ÏóêÌîºÏÜåÎìú
    total_computations = 0  # Ï¥ù Q ÏóÖÎç∞Ïù¥Ìä∏ ÌöüÏàò

    for i in tqdm(range(num_episodes)):

        # episode
        episode = []
    
        # ÌôòÍ≤Ω Ï¥àÍ∏∞Ìôî
        state, _ = env.reset()
    
        # reward Ïùò Ìï©ÏùÑ 0ÏúºÎ°ú Ï¥àÍ∏∞Ìôî
        reward_sum = 0
    
        # ÏóêÌîºÏÜåÎìúÍ∞Ä ÎÅùÎÇ¨ÏùåÏùÑ ÏïåÎ¶¨Îäî done_maskÎ•º FalseÎ°ú Ï¥àÍ∏∞Ìôî
        done = False
    
        # ÎûúÎç§ ÎÖ∏Ïù¥Ï¶à ÏûëÏÑ±.Ï≤òÏùåÏóî 1Ïùò Í∞íÏóêÏÑú, ÏóêÌîºÏÜåÎìú 10ÎßàÎã§ Í∞íÏù¥ Ï†êÏ∞® Ï§ÑÏñ¥Îì§ÎèÑÎ°ù ÏûëÏÑ±
        decay_ratio = 1./((i//10) + 1) 
        random_noise = np.random.randn(1, env.action_space.n) * decay_ratio

        #Ïù¥ Îïå, epsilon Í∞íÏùÄ 50 episodeÎßàÎã§ Í∞êÏÜåÌïòÎèÑÎ°ù ÏÑ§Ï†ï
        e = 1. / ((i//50) + 1)
    
        action = np.argmax(Q[state, :] + random_noise)

        eligibility_traces = np.zeros_like(Q)
        
        while not done:
            new_state, reward, done, truncated, info = env.step(action)
    
            #QÍ∞íÏóê ÎûúÎç§ ÎÖ∏Ïù¥Ï¶àÎ•º Ï∂îÍ∞ÄÌïú Í∞íÎì§ Ï§ë, Í∞ÄÏû• ÌÅ∞ Í∞íÏùÑ ÏÑ†ÌÉùÌïòÎèÑÎ°ù ÏûëÏÑ±
            #epsilonÏùò ÌôïÎ•†Î°ú Î¨¥ÏûëÏúÑ ÌñâÎèôÏùÑ ÏÑ†ÌÉù
            if np.random.rand(1) < e:
                new_action = env.action_space.sample()
            #Ïù¥Ïô∏Ïùò Í≤ΩÏö∞Ïóî QÍ∞íÏùÑ Îî∞ÎùºÍ∞ÄÎèÑÎ°ù ÌñâÎèôÏùÑ ÏÑ†ÌÉù
            else:
                new_action = np.argmax(Q[new_state, :] + random_noise)
    
            # ‚úÖ TD Error Í≥ÑÏÇ∞
            next_value = 0 if done else Q[new_state, new_action]
            td_error = reward + gamma * next_value - Q[state, action]

            # ‚úÖ Eligibility Traces ÏóÖÎç∞Ïù¥Ìä∏
            eligibility_traces[state, action] += 1  # Î∞©Î¨∏Ìïú ÏÉÅÌÉú-ÌñâÎèô ÏåçÎßå Ï¶ùÍ∞Ä

            # ‚úÖ Q ÌÖåÏù¥Î∏î ÏóÖÎç∞Ïù¥Ìä∏ (Î∞©Î¨∏Ìïú ÏÉÅÌÉú-ÌñâÎèôÎßå ÏóÖÎç∞Ïù¥Ìä∏)
            Q += alpha * td_error * eligibility_traces

            # ‚úÖ Eligibility Traces Decay (Œª Ï†ÅÏö©)
            eligibility_traces *= gamma * lam

            reward_sum += reward
            state = new_state
            action = new_action

        total_reward.append(reward_sum)
        total_reward_sum.append(np.mean(total_reward))

        # ‚úÖ Best Score Í∏∞Î°ù
        if reward_sum > best_score:
            best_score = reward_sum
            best_episode = i + 1  # 0-indexÍ∞Ä ÏïÑÎãàÎùº 1-indexÎ°ú Í∏∞Î°ù
            execution_time = time.time() - start_time
            break
    
    return Q, total_reward, total_reward_sum, execution_time, best_score, best_episode
```

```python
# Q-learning with Noise
def cliff_qlearning(env, gamma, num_episodes, alpha):

    Q = np.zeros([env.observation_space.n, env.action_space.n])

    # ‚úÖ Ïã§Ìñâ ÏãúÍ∞Ñ Ï∏°Ï†ï ÏãúÏûë
    execution_time = 0.0
    start_time = time.time()

    
    total_reward = []
    total_reward_sum = []

    best_score = -13  # ÌòÑÏû¨ÍπåÏßÄÏùò ÏµúÍ≥† Ï†êÏàò
    best_episode = -1  # ÏµúÍ≥† Ï†êÏàòÎ•º Îã¨ÏÑ±Ìïú ÏóêÌîºÏÜåÎìú
    total_computations = 0  # Ï¥ù Q ÏóÖÎç∞Ïù¥Ìä∏ ÌöüÏàò
    
    for i in tqdm(range(num_episodes)):
    
        # ÌôòÍ≤Ω Ï¥àÍ∏∞Ìôî
        state, _ = env.reset()
    
        # reward Ïùò Ìï©ÏùÑ 0ÏúºÎ°ú Ï¥àÍ∏∞Ìôî
        reward_sum = 0
    
        # ÏóêÌîºÏÜåÎìúÍ∞Ä ÎÅùÎÇ¨ÏùåÏùÑ ÏïåÎ¶¨Îäî done_maskÎ•º FalseÎ°ú Ï¥àÍ∏∞Ìôî
        done = False
    
        # ÎûúÎç§ ÎÖ∏Ïù¥Ï¶à ÏûëÏÑ±.Ï≤òÏùåÏóî 1Ïùò Í∞íÏóêÏÑú, ÏóêÌîºÏÜåÎìú 10ÎßàÎã§ Í∞íÏù¥ Ï†êÏ∞® Ï§ÑÏñ¥Îì§ÎèÑÎ°ù ÏûëÏÑ±
        decay_ratio = 1./((i//10) + 1) 
        random_noise = np.random.randn(1, env.action_space.n) * decay_ratio

        #Ïù¥ Îïå, epsilon Í∞íÏùÄ 50 episodeÎßàÎã§ Í∞êÏÜåÌïòÎèÑÎ°ù ÏÑ§Ï†ï
        e = 1. / ((i//50) + 1)
    
        while not done:

            #epsilonÏùò ÌôïÎ•†Î°ú Î¨¥ÏûëÏúÑ ÌñâÎèôÏùÑ ÏÑ†ÌÉù
            if np.random.rand(1) < e:
                action = env.action_space.sample()
            #Ïù¥Ïô∏Ïùò Í≤ΩÏö∞Ïóî QÍ∞íÏùÑ Îî∞ÎùºÍ∞ÄÎèÑÎ°ù ÌñâÎèôÏùÑ ÏÑ†ÌÉù
            else:
                action = np.argmax(Q[state, :] + random_noise)
                
            new_state, reward, done, truncated, info = env.step(action)
    
            #QÍ∞íÏóê ÎûúÎç§ ÎÖ∏Ïù¥Ï¶àÎ•º Ï∂îÍ∞ÄÌïú Í∞íÎì§ Ï§ë, Í∞ÄÏû• ÌÅ∞ Í∞íÏùÑ ÏÑ†ÌÉùÌïòÎèÑÎ°ù ÏûëÏÑ±
            new_action = np.argmax(Q[new_state, :] + random_noise)
    
            #SARSA ÏóÖÎç∞Ïù¥Ìä∏ Î∞©Î≤ïÏúºÎ°ú Q ÌÖåÏù¥Î∏î ÏóÖÎç∞Ïù¥Ìä∏ ÏΩîÎìúÎ•º ÏûëÏÑ±
            Q[state, action] = Q[state, action] + alpha * (reward + gamma * np.max(Q[new_state, :]) - Q[state, action])
    
            reward_sum += reward
            state = new_state
    
        total_reward.append(reward_sum)
        total_reward_sum.append(sum(total_reward)/(i+1))

        # ‚úÖ Best Score Í∏∞Î°ù
        if reward_sum > best_score:
            best_score = reward_sum
            best_episode = i + 1  # 0-indexÍ∞Ä ÏïÑÎãàÎùº 1-indexÎ°ú Í∏∞Î°ù
            execution_time = time.time() - start_time
            break
    
    return Q, total_reward, total_reward_sum, execution_time, best_score, best_episode
```

```python
# ÌôòÍ≤Ω Î∞è ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
env = gym.make("CliffWalking-v0")
discount_factor = 0.99
num_episodes = 300
alpha = 0.9
```

```python
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(8, 6))

m_list = ['sarsa', 'sarsa-lambda', 'q-learning']
best_s = []
t = []
max_episodes = 0  # ÏµúÎåÄ ÏóêÌîºÏÜåÎìú Ïàò Ï†ÄÏû•

plt.figure(figsize=(10, 6))  # Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÏÑ§Ï†ï

for idx, model in enumerate(m_list):
    if model == 'sarsa':
        Q, total_reward, total_reward_sum, execution_time, best_score, best_episode = cliff_sarsa(env, discount_factor, num_episodes, alpha)
    elif model == 'sarsa-lambda':
        Q, total_reward, total_reward_sum, execution_time, best_score, best_episode = cliff_sarsa_lambda(env, discount_factor, num_episodes, alpha)
    else:
        Q, total_reward, total_reward_sum, execution_time, best_score, best_episode = cliff_qlearning(env, discount_factor, num_episodes, alpha)

    best_s.append(best_score)
    t.append(execution_time)
    window = make_plot_list(total_reward)

    # ÏµúÎåÄ ÏóêÌîºÏÜåÎìú ÏóÖÎç∞Ïù¥Ìä∏ (Î™®Îì† Î™®Îç∏ÏùÑ Í∞ôÏùÄ xÏ∂ïÏóê ÎßûÏ∂îÍ∏∞ ÏúÑÌï¥)
    max_episodes = max(max_episodes, len(window))

    # Í∑∏ÎûòÌîÑ ÌîåÎ°Ø
    plt.plot(range(len(window)), window, linestyle='-', label=f"[{model}] Best Ep: {best_episode}")

    # üìå Best Score ÏßÄÏ†êÏóê 'Best!' ÌëúÏãú
    plt.axvline(best_episode, color='gray', linestyle='--', alpha=0.7)  # ÏàòÏßÅÏÑ† Ï∂îÍ∞Ä
    plt.text(best_episode, max(window) * 0.9, "Best!", fontsize=12, color='red', ha='center', weight='bold')

# üìå xÏ∂ï Í∏∏Ïù¥Î•º Í∞ÄÏû• Í∏¥ Î™®Îç∏ Í∏∞Ï§ÄÏúºÎ°ú ÎßûÏ∂§
plt.xlim(0, max_episodes)

plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from gym import spaces

class CustomCliffWalkingEnv(gym.Env):
    """
    CliffWalking-v0ÏùÑ Í∏∞Î∞òÏúºÎ°ú Ìïú Ïª§Ïä§ÌÖÄ ÌôòÍ≤Ω.
    - Í∏∞Ï°¥ ÌÅ¥Î¶¨ÌîÑÏóê Ï∂îÍ∞ÄÎ°ú "Íµ¨Î©ç(Holes)"ÏùÑ Ï§ëÍ∞ÑÏóê Î∞∞Ïπò.
    - Íµ¨Î©çÏóê ÎèÑÎã¨ÌïòÎ©¥ ÏãúÏûëÏ†êÏúºÎ°ú ÎêòÎèåÏïÑÍ∞ÄÍ≥† Î≥¥ÏÉÅÏùÑ ÏÉàÎ°≠Í≤å ÏÑ§Ï†ï.
    """

    def __init__(self):
        super(CustomCliffWalkingEnv, self).__init__()

        self.shape = (4, 12)  # Í∏∞Ï°¥ ÌÅ¨Í∏∞ Ïú†ÏßÄ
        self.start_state = (3, 0)  # ÏãúÏûë ÏúÑÏπò
        self.goal_state = (3, 11)  # Î™©Ìëú ÏúÑÏπò
        self.cliff = [(3, i) for i in range(1, 11)]  # Í∏∞Ï°¥ Cliff
        self.holes = [(0, 8), (1, 10)]  # ‚úÖ Ï∂îÍ∞ÄÎêú "Íµ¨Î©ç(Holes)" Ï¢åÌëú (Ïó¨Í∏∞ ÎèÑÎã¨ÌïòÎ©¥ Î¶¨ÏÖã)
        
        self.observation_space = spaces.Discrete(self.shape[0] * self.shape[1])
        self.action_space = spaces.Discrete(4)  # 0: ÏúÑ, 1: ÏïÑÎûò, 2: Ïò§Î•∏Ï™Ω, 3: ÏôºÏ™Ω

        self.reset()

    def reset(self):
        self.state = self.start_state  # ÏãúÏûëÏ†êÏúºÎ°ú Î¶¨ÏÖã
        return self.get_state_index(), {}

    def step(self, action):
        row, col = self.state

        # ‚úÖ ÌñâÎèô Ï†ïÏùò (Up, Down, Right, Left)
        if action == 0 and row > 0:  # ÏúÑÎ°ú Ïù¥Îèô
            row -= 1
        elif action == 1 and row < self.shape[0] - 1:  # ÏïÑÎûòÎ°ú Ïù¥Îèô
            row += 1
        elif action == 2 and col < self.shape[1] - 1:  # Ïò§Î•∏Ï™Ω Ïù¥Îèô
            col += 1
        elif action == 3 and col > 0:  # ÏôºÏ™Ω Ïù¥Îèô
            col -= 1

        self.state = (row, col)

        # ‚úÖ Î≥¥ÏÉÅ & Ï¢ÖÎ£å Ï°∞Í±¥ ÏÑ§Ï†ï
        if self.state == self.goal_state:
            reward = 100  # Î™©Ìëú ÎèÑÏ∞©
            done = True
        elif self.state in self.cliff:
            reward = -100  # CliffÏóê Îñ®Ïñ¥Ïßê
            self.state = self.start_state  # ÏãúÏûëÏ†êÏúºÎ°ú Î¶¨ÏÖã
            done = False
        elif self.state in self.holes:
            reward = -50  # Íµ¨Î©çÏóê Îπ†Ïßê
            self.state = self.start_state  # ÏãúÏûëÏ†êÏúºÎ°ú Î¶¨ÏÖã
            done = False
        else:
            reward = -2  # ÏùºÎ∞ò Ïù¥Îèô (Îçî Ïñ¥Î†µÍ≤å ÏÑ§Ï†ï)
            done = False

        return self.get_state_index(), reward, done, False, {}

    def get_state_index(self):
        """2D Ï¢åÌëúÎ•º 1D ÏÉÅÌÉú Ïù∏Îç±Ïä§Î°ú Î≥ÄÌôò"""
        return self.state[0] * self.shape[1] + self.state[1]
```

```python
# ÏÉàÎ°úÏö¥ ÌôòÍ≤Ω ÏÉùÏÑ±
env = CustomCliffWalkingEnv()
discount_factor = 0.99
num_episodes = 300
alpha = 0.9
```

```python
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(8, 6))

m_list = ['sarsa', 'sarsa-lambda', 'q-learning']
best_s = []
t = []
max_episodes = 0  # ÏµúÎåÄ ÏóêÌîºÏÜåÎìú Ïàò Ï†ÄÏû•

plt.figure(figsize=(10, 6))  # Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÏÑ§Ï†ï

for idx, model in enumerate(m_list):
    if model == 'sarsa':
        Q, total_reward, total_reward_sum, execution_time, best_score, best_episode = cliff_sarsa(env, discount_factor, num_episodes, alpha)
    elif model == 'sarsa-lambda':
        Q, total_reward, total_reward_sum, execution_time, best_score, best_episode = cliff_sarsa_lambda(env, discount_factor, num_episodes, alpha)
    else:
        Q, total_reward, total_reward_sum, execution_time, best_score, best_episode = cliff_qlearning(env, discount_factor, num_episodes, alpha)

    best_s.append(best_score)
    t.append(execution_time)
    window = make_plot_list(total_reward)

    # ÏµúÎåÄ ÏóêÌîºÏÜåÎìú ÏóÖÎç∞Ïù¥Ìä∏ (Î™®Îì† Î™®Îç∏ÏùÑ Í∞ôÏùÄ xÏ∂ïÏóê ÎßûÏ∂îÍ∏∞ ÏúÑÌï¥)
    max_episodes = max(max_episodes, len(window))

    # Í∑∏ÎûòÌîÑ ÌîåÎ°Ø
    plt.plot(range(len(window)), window, linestyle='-', label=f"[{model}] Best Ep: {best_episode}")

    # üìå Best Score ÏßÄÏ†êÏóê 'Best!' ÌëúÏãú
    plt.axvline(best_episode, color='gray', linestyle='--', alpha=0.7)  # ÏàòÏßÅÏÑ† Ï∂îÍ∞Ä
    plt.text(best_episode, max(window) * 0.9, "Best!", fontsize=12, color='red', ha='center', weight='bold')

# üìå xÏ∂ï Í∏∏Ïù¥Î•º Í∞ÄÏû• Í∏¥ Î™®Îç∏ Í∏∞Ï§ÄÏúºÎ°ú ÎßûÏ∂§
plt.xlim(0, max_episodes)

plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.legend()
plt.grid(True)
plt.show()
```
